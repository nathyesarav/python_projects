# Distancias

Medir la distancia entre dos datos es un proceso muy importante en Data Science.

![](distancia.jpg)

Normalmente visualizamos los datos como puntos o vectores en un plano cuando hablamos de sus distancias. Sin embargo es posible trabajar con distancias entre datos más complejos, como por ejemplo palabras.

Hay distintas formas de medir las distancias, cada una con sus ventajas, inconvenientes y casos más apropiados de uso:

- Distancia euclídea (o error cuadrático): La distancia euclídea es una de las más comunes y se basa en la idea de la distancia entre dos puntos en un plano. Se calcula como la raíz cuadrada de la suma de las diferencias de cada componente del vector al cuadrado. Es una medida de distancia muy comúnmente utilizada en el análisis de datos, ya que se basa en la intuición de la distancia "a pie" entre dos puntos.

- Distancia Manhattan (Taxicab geometry): La distancia de Manhattan, también conocida como la distancia "taxi", se calcula como la suma absoluta de las diferencias en cada componente del vector. Esta medida de distancia se utiliza a menudo cuando se trabaja con datos categóricos, ya que no requiere que los datos estén en una escala numérica.

- Distancia Chevyshev: La distancia de Chebyshev es una medida de similitud entre dos puntos en un espacio n-dimensional. Se calcula como la distancia máxima entre los valores de cada dimensión de los dos puntos. Por ejemplo, en un espacio bidimensional, la distancia de Chebyshev entre dos puntos (x1, y1) y (x2, y2) se calcula como max(|x1-x2|, |y1-y2|). Esta distancia se utiliza a menudo en problemas de clasificación y clustering, ya que es robusta a outliers y es menos sensible a la escala de los datos que otras medidas de distancia como la distancia euclidiana.

![](ChebyshevDistance.jpg)

- Distancia de Levenshtein: La distancia de Levenshtein es una medida de distancia entre dos cadenas de caracteres, y se calcula como el número mínimo de operaciones necesarias para convertir una cadena en otra. Esta medida de distancia se utiliza a menudo en el procesamiento del lenguaje natural y la comparación de texto.

![](levenshteinDistance.png)

````python
# Función en Python que calcula la distancia Manhattan entre dos puntos de un plano
def distance_manhattan(point1, point2):
  # Obtener las coordenadas de cada punto
  x1, y1 = point1
  x2, y2 = point2
  
  # Calcular la distancia de Manhattan
  distance = abs(x1 - x2) + abs(y1 - y2)
  
  return distance
````

````python
# Para utilizar esta función, pásale dos tuplas que representen a cada punto. Por ejemplo:
point1 = (1, 2)
point2 = (4, 5)

distance = distance_manhattan(point1, point2)
print(distance)
````

## ¿Qué papel juega el error cuadrático en cross-validation?

El error cuadrático es una forma de medir la distancia euclídea entre el vector predicción y el vector dato real.

En el proceso de validación cruzada, el error cuadrático medio (MSE, por sus siglas en inglés) se utiliza a menudo como una medida de rendimiento para evaluar el rendimiento de un modelo. El MSE se calcula como la suma de los errores al cuadrado entre las predicciones del modelo y los valores observados, divididos por el número de datos de entrenamiento.

![](crossValidation-overfitting-underfitting.png)

La validación cruzada es una técnica utilizada para evaluar el rendimiento de un modelo de aprendizaje automático. Consiste en dividir los datos de entrenamiento en un conjunto de prueba y un conjunto de entrenamiento y entrenar el modelo con el conjunto de entrenamiento. Luego, se evalúa el rendimiento del modelo utilizando el conjunto de prueba. Este proceso se repite varias veces, utilizando diferentes divisiones del conjunto de datos para entrenar y evaluar el modelo.

El MSE se utiliza a menudo como una medida de rendimiento en la validación cruzada porque es una medida de la diferencia entre las predicciones del modelo y los valores observados. El MSE más bajo indica un mejor rendimiento del modelo, ya que significa que las predicciones del modelo están más cerca de los valores observados. Sin embargo, el MSE no es la única medida de rendimiento que se puede utilizar en la validación cruzada. Otros métricas comunes incluyen la precisión, el puntaje F1 y el área bajo la curva ROC (AUC).

### Sobreajuste y Subajuste

Como la validación cruzada se realiza dividiendo el conjunto de datos en varios subconjuntos, llamados **"folds" (pliegues)**, y entrenando el modelo en diferentes combinaciones de estos subconjuntos, con uno de ellos utilizado como conjunto de prueba, se pueden encontrar dos casos:

- **Sobreajuste (overfitting)** se refiere a un modelo que se ajusta demasiado bien a los datos de entrenamiento, pero no generaliza bien a datos nuevos. Esto sucede cuando un modelo tiene demasiados parámetros o es demasiado complejo para el tamaño del conjunto de datos de entrenamiento.

- El **subajuste (underfitting)** se refiere a un modelo que no se ajusta lo suficientemente bien a los datos de entrenamiento. Esto sucede cuando un modelo es demasiado simple o tiene demasiado pocos parámetros para capturar la complejidad de los datos de entrenamiento.

![](Training-error-and-cross-validation-error-as-a-function-of-model-complexity.png)

**La validación cruzada se utiliza para evitar el sobreajuste y el subajuste**, ya que permite evaluar el rendimiento del modelo en diferentes subconjuntos de datos. Si un modelo tiene un rendimiento similar en todos los pliegues, es probable que sea un buen modelo para generalizar a nuevos datos. Sin embargo, si el rendimiento del modelo varía significativamente entre los pliegues, puede indicar que el modelo está sobreajustado o subajustado.

### Funciones loss en entrenamiento de redes neuronales

Recordemos estas líneas de código para compilar una red neuronal en Keras:

```
# Compilamos el modelo con una tasa de aprendizaje y una función de pérdida
model.compile(optimizer=tf.keras.optimizers.Adam(1), loss='mean_squared_error')
```

La función compile() se utiliza para configurar el proceso de entrenamiento del modelo.

El primer argumento es el **optimizador**, en este caso se esta utilizando Adam, que es un optimizador de gradiente estocástico, es uno de los optimizadores más populares y utilizado para modelos de aprendizaje automático. El número 1 es la tasa de aprendizaje del optimizador.

El segundo argumento es la **función de pérdida**, en este caso se esta utilizando la "mean_squared_error", que es el error cuadrático medio. Esta función de pérdida se utiliza a menudo para problemas de regresión, ya que mide la diferencia entre los valores previstos y los valores reales.

Veamos otras funciones de pérdida posibles:

- **Binary Cross-Entropy**: se utiliza para problemas de clasificación binaria. Mide la diferencia entre la distribución de probabilidad prevista y la verdadera distribución de probabilidad para cada clase.

- **Categorical Cross-Entropy**: se utiliza para problemas de clasificación multiclase. Es similar a la entropía cruzada binaria, pero se utiliza para problemas con más de dos clases.

- **Sparse Categorical Cross-Entropy**: Similar a la Categorical Cross-Entropy pero se utiliza cuando las etiquetas son enteros en lugar de vectores one-hot.

- **Mean Absolute Error (MAE)**: Se utiliza para problemas de regresión. Mide la diferencia entre los valores previstos y los valores reales en términos de valor absoluto.

- **Mean Squared Error (MSE)**: Similar al MAE, pero mide la diferencia en términos del cuadrado de los valores.

- **Huber Loss**: Es una mezcla entre la entropía cruzada binaria y el MSE, que es menos sensible a outliers.

- **Kullback-Leibler divergence**: Utilizada en problemas de clasificación y selección de características. Es una medida de la diferencia entre dos distribuciones de probabilidad.

Cada función de pérdida tiene sus propias propiedades y se utiliza en diferentes tipos de problemas, por lo que es importante elegir la función de pérdida adecuada para el problema específico que se está tratando.

![](lossFunctionNeuralNetwork.png)

Una función de pérdida, también conocida como **función de costo** o **función objetivo**, es una medida numérica que indica qué tan lejos están los resultados previstos del modelo de los resultados reales. La función de pérdida se utiliza para evaluar la precisión del modelo y para guiar el proceso de entrenamiento. El objetivo es minimizar la función de pérdida para mejorar la precisión del modelo.

La **distancia** o el **error** son términos generales que se refieren a la diferencia entre dos valores o dos conjuntos de valores. En el contexto del aprendizaje automático, también se puede utilizar el término "distancia" o "error" para referirse a la diferencia entre los resultados previstos del modelo y los resultados reales.

# Levenshtein, distancia en textos

````python
# Función en Python que calcula la distancia de Levenshtein entre dos palabras:
# La distancia de Levenshtein es una medida de distancia entre dos cadenas de
# caracteres y se calcula como el número mínimo de operaciones necesarias para
# convertir una cadena en otra. 
def distance_levenshtein(word1, word2):
  # Obtener la longitud de cada palabra
  len1 = len(word1)
  len2 = len(word2)
  
  # Inicializar una matriz de distancias con ceros
  distance = [[0] * (len2 + 1) for _ in range(len1 + 1)]
  
  # Asignar costes a las operaciones de eliminación, inserción y sustitución
  cost_delete = 1
  cost_insert = 1
  cost_replace = 1
  
  # Rellenar la primera fila y columna con los valores de 0 a len2
  for i in range(len2 + 1):
    distance[0][i] = i
  for i in range(len1 + 1):
    distance[i][0] = i
    
  # Iterar a través de las filas y columnas de la matriz de distancias
  for i in range(1, len1 + 1):
    for j in range(1, len2 + 1):
      # Calcular el coste de cada operación
      delete = distance[i - 1][j] + cost_delete
      insert = distance[i][j - 1] + cost_insert
      replace = distance[i - 1][j - 1]
      if word1[i - 1] != word2[j - 1]:
        replace += cost_replace
      
      # Tomar el menor coste entre eliminación, inserción y sustitución
      distance[i][j] = min(delete, insert, replace)
  
  # Devolver la última celda de la matriz de distancias como la distancia de Levenshtein
  return distance[len1][len2]
````

````python
# Para utilizar esta función, simplemente pásale dos cadenas de caracteres que representen a las palabras
word1 = "lección"
word2 = "acción"

distance = distance_levenshtein(word1, word2)
print(distance)  # Imprime 3
````

````python
# En Python también puedes utilizar la función distance.levenshtein del módulo
# distance del paquete python-Levenshtein para calcular la distancia
# de Levenshtein entre dos cadenas de caracteres.
#
# Para utilizar esta función quizá debes instalar el paquete python-Levenshtein
# utilizando pip, que es el gestor de paquetes de Python:
# pip3 install python-Levenshtein
#
# Una vez instalado, puedes importar la función Levenshtein.distance
# del módulo rapidfuzz.distance y utilizarla de la siguiente manera:

from rapidfuzz.distance import Levenshtein

word1 = "lección"
word2 = "acción"

distance = Levenshtein.distance(word1, word2)
print(distance)
````

# Otras distancias en Data Science

- **Distancia de Jaccard**: mide la similitud entre dos conjuntos de elementos, se calcula como la relación entre el número de elementos en común y el número total de elementos únicos en ambos conjuntos.

- **Distancia de Sørensen-Dice**: se utiliza para medir la similitud entre dos conjuntos de elementos. Es una variante de la distancia de Jaccard y se calcula como el cociente entre el número de elementos en común y el número de elementos únicos entre los dos conjuntos. En otras palabras, se calcula como 2 * (intersección de los conjuntos) / (tamaño del conjunto A + tamaño del conjunto B). La ventaja de la distancia de Sørensen-Dice en comparación con la distancia de Jaccard es que tiene en cuenta el tamaño relativo de los conjuntos. La distancia de Sørensen-Dice es comúnmente utilizada en la comparación de conjuntos de datos en tareas de **procesamiento de lenguaje natural** (NLP), como el procesamiento de texto, la comparación de documentos y la búsqueda de similitudes.

- **Distancia de Hamming**: se utiliza para medir la diferencia entre dos cadenas de caracteres de la misma longitud. Se calcula como el número de posiciones en las que los caracteres son diferentes.

- **Distancia del coseno**: se utiliza para medir la similitud entre dos vectores. Se calcula como el coseno del ángulo entre los vectores.

- **Distancia de Mahalanobis**: es una medida de distancia multivariable que tiene en cuenta la **covarianza** entre las variables.

# Covarianza

La covarianza es una herramienta estadística usada para medir la relación entre dos variables. Representa la variación conjunta de dos variables y nos ayuda a determinar si se mueven juntas o no. La covarianza se puede calcular utilizando la siguiente fórmula:

covariance = (E[XY] - E[X]E[Y]) / n

Donde E[XY] es la media de xy, E[X] la media de X, E[Y] la media de Y, y n es el número de observaciones. La covarianza puede tener valores entre -∞ y +∞, donde un valor cercano a 0 indica que las variables no están correlacionadas, un valor positivo indica una correlación positiva, y un valor negativo indica una correlación negativa. La covarianza se usa comúnmente para unir los índices de dos mercados y calcular la correlación entre diferentes tipos de activos financieros.

![](positive_negative_weak_covariance.png)

En **aprendizaje automático** se utiliza en algoritmos de clasificación y regresión para medir la relación entre las **variables predictoras** y la **variable objetivo**.
