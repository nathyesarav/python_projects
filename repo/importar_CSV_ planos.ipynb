La importación se logró sin problemas por el formato en el que viene el archivo (valores separados por comas) pero ¿que sucede si el separador no son comas?

```python
spotifyDataNoComa = pd.read_csv("spotify_data_notcoma.csv")
```

```python
spotifyDataNoComa.head()
```

Parametro SEP para separar 

```python
spotifyDataNoComa = pd.read_csv("spotify_data_notcoma.csv", sep=";")
spotifyDataNoComa.head()
```

Agregamos a este archivo decimales ¿fueron reconocidos como números?

```python
spotifyDataNoComa.dtypes
```

Parámetro Decimal
Con decimal y thousands puedes definir la importación de números. 

```python
spotifyDataNoComa = pd.read_csv("spotify_data_notcoma.csv", sep=";", decimal=",")
spotifyDataNoComa.head()
```

```python
spotifyDataNoComa.dtypes
```

Parámetro Name
¿te gustan los títulos de las columnas o los prefieres en español?

```python
spotifyDataNoComa = pd.read_csv("spotify_data_notcoma.csv", header=0, sep=";", decimal=",", names=["nombreArtista", "nombreCancion",
                                                                                         "CancionID", "popularidad"])
spotifyDataNoComa.head()
```

Vamos a importar un nuevo archivo, en este caso unos resultados de una campaña de marketing.

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html

Nos encontramos de nuevo con el separador tabular.  Recuerda que la combinación \t, \n o similares son conocidos como caracteres de escape y tiene su interpretación en el lenguaje de programación

```python
marketing_campaign = pd.read_csv("marketing_campaign.csv", sep="\t")
```

```python
marketing_campaign.head()
```

Los cambios en el separador nos permitieron cargar los datos, pero tenemos números y fechas ¿se importaron de forma correcta?

```python
marketing_campaign.dtypes
```

ID puede ser un campo Clave por lo que aquellos suceptibles de relacionarse con otras abses es mejor tenerlos en formato texto o String. ¿Qué opinas de las fechas?

```python
marketing_campaign['ID'] =  marketing_campaign['ID'].map(str)
marketing_campaign['Dt_Customer'] =  pd.to_datetime(marketing_campaign['Dt_Customer'])
marketing_campaign.head()
```

```python
marketing_campaign.dtypes
```
