```python
import re
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud 
plt.rcParams.update({'figure.figsize': (8, 6), 'figure.dpi': 120})
```

```python
file = open("test.txt", encoding="UTF-8")
data = file.read()
textos = data.split('\nQ')
```

```python
textos = [re.sub(r"^\d+\s", "", elemento) for elemento in textos]
preguntas = []
for texto in textos:
    texto = preguntas.append(texto[:texto.find('?')])     
preguntas      
```

```python
df = pd.DataFrame(preguntas)
df.rename(columns={0 :'pregunta'}, inplace=True)
df['pregunta'] = df['pregunta'].apply(lambda x: x[1:])
df
```

```python
titles=''.join(str(df.pregunta)).lower()
titles
```

```python
plt.figure(figsize=(8,6))
wordcloud = WordCloud(collocations=True, max_words=500 ).generate(titles)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
