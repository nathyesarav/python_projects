### pdist y squareform de sciPy

Squareform es una función que se utiliza para dar formato a una matriz de distancia. Se suele emplear para convertir la salida de la función 'pdist' en una matriz de distancia cuadrada. Veamos unos ejemplos de cómo se usa la función:

```python
import numpy as np
from scipy.spatial.distance import pdist, squareform

# Create a random matrix of 2x2
X = np.random.rand(2, 2)
X
```

```python
X[:, 0]
```

```python
X[:, 1]
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1])
```

```python
# Compute the Euclidean distance between all pairs of points
dist_mat = pdist(X)
dist_mat
```

```python
# Convert the distance matrix to squareform
square_dist_mat = squareform(dist_mat)
square_dist_mat
```

En Python, la función squareform() se utiliza para convertir una matriz de distancia condensada en una matriz de distancia completa.

La matriz de distancia condensada es un arreglo unidimensional que almacena la matriz de distancia completa en forma condensada, es decir, solo almacena las distancias únicas, ya que la diagonal siempre es cero. La función squareform() convierte esta matriz condensada en una matriz de distancia completa cuadrada.

Nota curiosa: En general, la cantidad de distancias distintas entre n puntos en un espacio de d dimensiones está dada por la siguiente fórmula:

C(n,2) = n(n-1)/2

donde C(n,2) es el número de combinaciones de n elementos tomados de 2 en 2 (es decir, el número de pares distintos que se pueden formar con n elementos).

Aplicando esta fórmula a nuestro caso, donde n=10 y d=10, obtenemos:

C(10,2) = 10(10-1)/2 = 45

Por lo tanto, hay 45 distancias distintas entre 10 puntos en un espacio de 10 dimensiones.

# Diferencia entre tupla y lista en Python

En Python, las tuplas y las listas son dos tipos de estructuras de datos que se utilizan para almacenar elementos en una secuencia ordenada. La principal diferencia entre ellas es que las tuplas son **inmutables**, es decir, una vez que se crea una tupla, **no se pueden modificar sus elementos**, mientras que las listas son mutables y se pueden modificar en cualquier momento.

Además de su inmutabilidad, otras diferencias entre las tuplas y las listas incluyen:

- Las tuplas se definen con paréntesis ( ) y las listas con corchetes [ ].

- Las tuplas son más eficientes en términos de espacio y tiempo de procesamiento.

- Las tuplas se utilizan generalmente para almacenar datos que no cambian, como las coordenadas de un punto en un plano, mientras que las listas se utilizan para almacenar colecciones de elementos que pueden ser añadidos, eliminados o modificados.

En resumen, si necesita una estructura de datos que no se pueda modificar después de su creación, use una tupla. Si necesita una estructura de datos que se pueda modificar, use una lista.


```python
# Crear una tupla
tupla1 = ('manzana', 'banana', 'naranja')

# Acceder a los elementos de una tupla por su índice
print(tupla1[0]) # 'manzana'
print(tupla1[1]) # 'banana'
```

```python
# Intentar modificar los elementos de una tupla da un error, ya que las tuplas son inmutables
tupla1[0] = 'pera' # TypeError: 'tuple' object does not support item assignment
```

```python
# Concatenar tuplas
tupla2 = ('sandía', 'melón')
tupla3 = tupla1 + tupla2
print(tupla3) # ('manzana', 'banana', 'naranja', 'sandía', 'melón')
```

```python
# Desempaquetar una tupla en variables
fruta1, fruta2, fruta3, fruta4, fruta5 = tupla3
print(fruta1) # 'manzana'
print(fruta2) # 'banana'
print(fruta3) # 'naranja'
print(fruta4) # 'sandía'
print(fruta5) # 'melón'
```

# Market basket analysis



El análisis de cesta de la compra o "market basket analysis" es una técnica de minería de datos que busca patrones en las compras de los clientes. Un ejemplo común de esta técnica se utiliza en la tienda minorista Amazon.com.

Imaginemos que una tienda minorista desea entender qué productos se compran juntos con mayor frecuencia. Para ello, podrían utilizar el análisis de cesta de la compra para analizar los datos de transacciones de los clientes y descubrir patrones interesantes.

Por ejemplo, supongamos que los datos de transacciones muestran que el 60% de los clientes que compran leche también compran pan y que el 50% de los clientes que compran pan también compran queso. La tienda minorista podría utilizar esta información para colocar estos productos juntos en el mismo pasillo o crear paquetes especiales que ofrezcan un descuento en la compra de estos productos juntos.

 

Además, la tienda minorista podría utilizar el análisis de cesta de la compra para identificar productos que se compran juntos con menos frecuencia, lo que podría ser una oportunidad para promocionar esos productos juntos para aumentar las ventas.

Para realizar un análisis de cesta de la compra en Python, se puede utilizar la librería "**apyori**", que incluye funciones para realizar análisis de cesta de la compra y reglas de asociación.

```python
# pip3 install apyori

from apyori import apriori
import pandas as pd

# Crear un conjunto de datos de ejemplo de transacciones de compra
datos = [['Leche', 'Pan', 'Queso'],
         ['Leche', 'Pan'],
         ['Leche', 'Queso'],
         ['Leche', 'Pan', 'Queso'],
         ['Leche', 'Pan', 'Queso', 'Huevos'],
         ['Leche', 'Pan', 'Huevos']]

# Utilizar el algoritmo Apriori para generar conjuntos de elementos frecuentes y reglas de asociación
conjuntos_frecuentes = list(apriori(datos, min_support=0.2, min_confidence=0.7, min_lift=1.2))

# Imprimir los conjuntos de elementos frecuentes y las reglas de asociación resultantes
print("Conjuntos Frecuentes:")
for conjunto in conjuntos_frecuentes:
    print(conjunto.items)
print("\nReglas de Asociación:")
for conjunto in conjuntos_frecuentes:
    for regla in conjunto.ordered_statistics:
        antecedente = list(regla.items_base)
        consecuente = list(regla.items_add)
        confianza = regla.confidence
        lift = regla.lift
        print(f"{antecedente} -> {consecuente}, confianza = {confianza:.2f}, lift = {lift:.2f}")
```

Se muestra los conjuntos de elementos frecuentes y las reglas de asociación más relevantes en función de su confianza y lift mínimo.

La confianza y el lift son dos medidas comunes utilizadas en el análisis de reglas de asociación, que se aplican para identificar patrones en conjuntos de datos de transacciones de compras.

La **confianza** es la medida de la frecuencia relativa con la que aparece un conjunto de elementos consecuentes en las transacciones que también contienen el conjunto de elementos antecedentes. Se calcula dividiendo el número de transacciones que contienen tanto el antecedente como el consecuente por el número total de transacciones que contienen el antecedente. La confianza se expresa como un porcentaje entre 0 y 100.

El **lift** es una medida de la importancia relativa de una regla de asociación en comparación con su frecuencia esperada en ausencia de cualquier relación entre los elementos antecedentes y consecuentes. El lift se calcula dividiendo la confianza de la regla por la frecuencia del consecuente en el conjunto de datos. Un valor de lift mayor que 1 indica una asociación positiva entre el antecedente y el consecuente, mientras que un valor menor que 1 indica una asociación negativa y un valor igual a 1 indica que no hay asociación.

# Métricas de pdist

- braycurtis: esta métrica se utiliza a menudo para comparar datos de composición, como los resultados de una encuesta en la que se pregunta a los participantes sobre la frecuencia con la que consumen diferentes alimentos. La distancia de Bray-Curtis es una medida de la diferencia entre dos perfiles de composición.

- canberra: esta métrica es adecuada para datos en los que los valores son relativamente pequeños y las diferencias en los valores más pequeños son más importantes que las diferencias en los valores más grandes.

- chebyshev: esta métrica es una medida de la máxima diferencia entre los valores de dos vectores.

- cityblock: esta métrica mide la distancia de Manhattan entre dos vectores, también llamada "taxicab".

- correlation: esta métrica mide la correlación entre dos vectores y se utiliza a menudo en análisis de series de tiempo y minería de datos.

- cosine: esta métrica mide la similitud del coseno entre dos vectores teniñendo en cuenta el ángulo entre ellos y no la magnitud de los mismos. Se utiliza a menudo en análisis de texto y minería de datos.

- dice: esta métrica se utiliza a menudo en análisis de texto y se basa en la cantidad de elementos que comparten dos conjuntos de elementos.

- euclidean: esta métrica mide la distancia euclidiana entre dos vectores.

- hamming: esta métrica mide la cantidad de elementos que difieren entre dos vectores.

- jaccard: esta métrica mide la similitud de Jaccard entre dos vectores y se utiliza a menudo en análisis de texto y minería de datos.

- jensenshannon: esta métrica mide la divergencia de Jensen-Shannon entre dos distribuciones de probabilidad.

- kulczynski1: esta métrica mide la cantidad de elementos que comparten dos conjuntos de elementos, pero con más peso en los elementos menos frecuentes.

- mahalanobis: esta métrica mide la distancia entre dos puntos teniendo en cuenta la covarianza de los datos.

- matching: esta métrica mide la cantidad de elementos que son iguales entre dos vectores.

- minkowski: esta métrica es una generalización de la distancia euclidiana y se puede utilizar para calcular diferentes tipos de distancias, como la distancia de Manhattan.

- rogerstanimoto: esta métrica mide la cantidad de elementos que son iguales entre dos vectores, pero con más peso en los elementos que están presentes en ambos vectores.

- russellrao: esta métrica mide la cantidad de elementos que son iguales entre dos vectores dividida por el número total de elementos.

- seuclidean: esta métrica es similar a la distancia euclidiana, pero tiene en cuenta la varianza de cada dimensión.

- sokalmichener: esta métrica mide la cantidad de elementos que son iguales entre dos vectores, pero con más peso en los elementos que están presentes en ambos vectores y menos peso en los elementos

# Visualizar matrices dispersas en Python

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.sparse import random

# Crear matriz dispersa aleatoria
n = 10
density = 0.3
matrix = random(n, n, density=density)
print(matrix.shape)
print("\n")
print(matrix)
```

```python
# Convertir matriz dispersa a una matriz densa
dense_matrix = matrix.todense()
print(dense_matrix)
```

```python
# Graficar matriz densa
plt.imshow(dense_matrix, cmap='gray', interpolation='nearest')
plt.show()
```

```python
# Graficar matriz dispersa
plt.spy(matrix, markersize=5)
plt.show()
```

matrix.todense() es un método de la clase scipy.sparse.csr_matrix que convierte una matriz dispersa en una matriz densa. En otras palabras, convierte una matriz representada de manera dispersa, en la que solo se almacenan los elementos distintos de cero, en una matriz en la que se almacenan todos los elementos.

El método todense() devuelve una matriz densa en formato de matriz de Numpy (clase numpy.matrix). Si la matriz dispersa original tenía forma (m, n), la matriz densa resultante tendrá forma (m, n) también. Todos los elementos de la matriz densa, tanto los elementos distintos de cero como los ceros, se almacenan en la memoria.

Es importante tener en cuenta que la conversión de una matriz dispersa a una matriz densa puede ser costosa en términos de uso de memoria y tiempo de ejecución, especialmente si la matriz dispersa es grande. Por lo tanto, en algunos casos, puede ser más eficiente trabajar con la matriz dispersa en su forma original y utilizar métodos y funciones específicas para matrices dispersas.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.sparse import random

# Crear matriz dispersa aleatoria
n = 1000
density = 0.1
matrix = random(n, n, density=density)

# Convertir matriz dispersa a una matriz densa
dense_matrix = matrix.todense()

# Graficar matriz densa
plt.imshow(dense_matrix, cmap='gray', interpolation='nearest')
plt.show()

# Graficar matriz dispersa
plt.spy(matrix, markersize=.05)
plt.show()
```
