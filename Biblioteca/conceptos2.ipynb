### pdist y squareform de sciPy

Squareform es una funci贸n que se utiliza para dar formato a una matriz de distancia. Se suele emplear para convertir la salida de la funci贸n 'pdist' en una matriz de distancia cuadrada. Veamos unos ejemplos de c贸mo se usa la funci贸n:

```python
import numpy as np
from scipy.spatial.distance import pdist, squareform

# Create a random matrix of 2x2
X = np.random.rand(2, 2)
X
```

```python
X[:, 0]
```

```python
X[:, 1]
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1])
```

```python
# Compute the Euclidean distance between all pairs of points
dist_mat = pdist(X)
dist_mat
```

```python
# Convert the distance matrix to squareform
square_dist_mat = squareform(dist_mat)
square_dist_mat
```

En Python, la funci贸n squareform() se utiliza para convertir una matriz de distancia condensada en una matriz de distancia completa.

La matriz de distancia condensada es un arreglo unidimensional que almacena la matriz de distancia completa en forma condensada, es decir, solo almacena las distancias 煤nicas, ya que la diagonal siempre es cero. La funci贸n squareform() convierte esta matriz condensada en una matriz de distancia completa cuadrada.

# Diferencia entre tupla y lista en Python

En Python, las tuplas y las listas son dos tipos de estructuras de datos que se utilizan para almacenar elementos en una secuencia ordenada. La principal diferencia entre ellas es que las tuplas son **inmutables**, es decir, una vez que se crea una tupla, **no se pueden modificar sus elementos**, mientras que las listas son mutables y se pueden modificar en cualquier momento.

Adem谩s de su inmutabilidad, otras diferencias entre las tuplas y las listas incluyen:

- Las tuplas se definen con par茅ntesis ( ) y las listas con corchetes [ ].

- Las tuplas son m谩s eficientes en t茅rminos de espacio y tiempo de procesamiento.

- Las tuplas se utilizan generalmente para almacenar datos que no cambian, como las coordenadas de un punto en un plano, mientras que las listas se utilizan para almacenar colecciones de elementos que pueden ser a帽adidos, eliminados o modificados.

En resumen, si necesita una estructura de datos que no se pueda modificar despu茅s de su creaci贸n, use una tupla. Si necesita una estructura de datos que se pueda modificar, use una lista.


```python
# Crear una tupla
tupla1 = ('manzana', 'banana', 'naranja')

# Acceder a los elementos de una tupla por su 铆ndice
print(tupla1[0]) # 'manzana'
print(tupla1[1]) # 'banana'
```

```python
# Intentar modificar los elementos de una tupla da un error, ya que las tuplas son inmutables
tupla1[0] = 'pera' # TypeError: 'tuple' object does not support item assignment
```

```python
# Concatenar tuplas
tupla2 = ('sand铆a', 'mel贸n')
tupla3 = tupla1 + tupla2
print(tupla3) # ('manzana', 'banana', 'naranja', 'sand铆a', 'mel贸n')
```

```python
# Desempaquetar una tupla en variables
fruta1, fruta2, fruta3, fruta4, fruta5 = tupla3
print(fruta1) # 'manzana'
print(fruta2) # 'banana'
print(fruta3) # 'naranja'
print(fruta4) # 'sand铆a'
print(fruta5) # 'mel贸n'
```

# Market basket analysis



El an谩lisis de cesta de la compra o "market basket analysis" es una t茅cnica de miner铆a de datos que busca patrones en las compras de los clientes. Un ejemplo com煤n de esta t茅cnica se utiliza en la tienda minorista Amazon.com.

Imaginemos que una tienda minorista desea entender qu茅 productos se compran juntos con mayor frecuencia. Para ello, podr铆an utilizar el an谩lisis de cesta de la compra para analizar los datos de transacciones de los clientes y descubrir patrones interesantes.

Por ejemplo, supongamos que los datos de transacciones muestran que el 60% de los clientes que compran leche tambi茅n compran pan y que el 50% de los clientes que compran pan tambi茅n compran queso. La tienda minorista podr铆a utilizar esta informaci贸n para colocar estos productos juntos en el mismo pasillo o crear paquetes especiales que ofrezcan un descuento en la compra de estos productos juntos.

 

Adem谩s, la tienda minorista podr铆a utilizar el an谩lisis de cesta de la compra para identificar productos que se compran juntos con menos frecuencia, lo que podr铆a ser una oportunidad para promocionar esos productos juntos para aumentar las ventas.

Para realizar un an谩lisis de cesta de la compra en Python, se puede utilizar la librer铆a "**apyori**", que incluye funciones para realizar an谩lisis de cesta de la compra y reglas de asociaci贸n.

```python
# pip3 install apyori

from apyori import apriori
import pandas as pd

# Crear un conjunto de datos de ejemplo de transacciones de compra
datos = [['Leche', 'Pan', 'Queso'],
         ['Leche', 'Pan'],
         ['Leche', 'Queso'],
         ['Leche', 'Pan', 'Queso'],
         ['Leche', 'Pan', 'Queso', 'Huevos'],
         ['Leche', 'Pan', 'Huevos']]

# Utilizar el algoritmo Apriori para generar conjuntos de elementos frecuentes y reglas de asociaci贸n
conjuntos_frecuentes = list(apriori(datos, min_support=0.2, min_confidence=0.7, min_lift=1.2))

# Imprimir los conjuntos de elementos frecuentes y las reglas de asociaci贸n resultantes
print("Conjuntos Frecuentes:")
for conjunto in conjuntos_frecuentes:
    print(conjunto.items)
print("\nReglas de Asociaci贸n:")
for conjunto in conjuntos_frecuentes:
    for regla in conjunto.ordered_statistics:
        antecedente = list(regla.items_base)
        consecuente = list(regla.items_add)
        confianza = regla.confidence
        lift = regla.lift
        print(f"{antecedente} -> {consecuente}, confianza = {confianza:.2f}, lift = {lift:.2f}")
```

Se muestra los conjuntos de elementos frecuentes y las reglas de asociaci贸n m谩s relevantes en funci贸n de su confianza y lift m铆nimo.

La confianza y el lift son dos medidas comunes utilizadas en el an谩lisis de reglas de asociaci贸n, que se aplican para identificar patrones en conjuntos de datos de transacciones de compras.

La **confianza** es la medida de la frecuencia relativa con la que aparece un conjunto de elementos consecuentes en las transacciones que tambi茅n contienen el conjunto de elementos antecedentes. Se calcula dividiendo el n煤mero de transacciones que contienen tanto el antecedente como el consecuente por el n煤mero total de transacciones que contienen el antecedente. La confianza se expresa como un porcentaje entre 0 y 100.

El **lift** es una medida de la importancia relativa de una regla de asociaci贸n en comparaci贸n con su frecuencia esperada en ausencia de cualquier relaci贸n entre los elementos antecedentes y consecuentes. El lift se calcula dividiendo la confianza de la regla por la frecuencia del consecuente en el conjunto de datos. Un valor de lift mayor que 1 indica una asociaci贸n positiva entre el antecedente y el consecuente, mientras que un valor menor que 1 indica una asociaci贸n negativa y un valor igual a 1 indica que no hay asociaci贸n.

# M茅tricas de pdist

- braycurtis: esta m茅trica se utiliza a menudo para comparar datos de composici贸n, como los resultados de una encuesta en la que se pregunta a los participantes sobre la frecuencia con la que consumen diferentes alimentos. La distancia de Bray-Curtis es una medida de la diferencia entre dos perfiles de composici贸n.

- canberra: esta m茅trica es adecuada para datos en los que los valores son relativamente peque帽os y las diferencias en los valores m谩s peque帽os son m谩s importantes que las diferencias en los valores m谩s grandes.

- chebyshev: esta m茅trica es una medida de la m谩xima diferencia entre los valores de dos vectores.

- cityblock: esta m茅trica mide la distancia de Manhattan entre dos vectores, tambi茅n llamada "taxicab".

- correlation: esta m茅trica mide la correlaci贸n entre dos vectores y se utiliza a menudo en an谩lisis de series de tiempo y miner铆a de datos.

- cosine: esta m茅trica mide la similitud del coseno entre dos vectores teni帽endo en cuenta el 谩ngulo entre ellos y no la magnitud de los mismos. Se utiliza a menudo en an谩lisis de texto y miner铆a de datos.

- dice: esta m茅trica se utiliza a menudo en an谩lisis de texto y se basa en la cantidad de elementos que comparten dos conjuntos de elementos.

- euclidean: esta m茅trica mide la distancia euclidiana entre dos vectores.

- hamming: esta m茅trica mide la cantidad de elementos que difieren entre dos vectores.

- jaccard: esta m茅trica mide la similitud de Jaccard entre dos vectores y se utiliza a menudo en an谩lisis de texto y miner铆a de datos.

- jensenshannon: esta m茅trica mide la divergencia de Jensen-Shannon entre dos distribuciones de probabilidad.

- kulczynski1: esta m茅trica mide la cantidad de elementos que comparten dos conjuntos de elementos, pero con m谩s peso en los elementos menos frecuentes.

- mahalanobis: esta m茅trica mide la distancia entre dos puntos teniendo en cuenta la covarianza de los datos.

- matching: esta m茅trica mide la cantidad de elementos que son iguales entre dos vectores.

- minkowski: esta m茅trica es una generalizaci贸n de la distancia euclidiana y se puede utilizar para calcular diferentes tipos de distancias, como la distancia de Manhattan.

- rogerstanimoto: esta m茅trica mide la cantidad de elementos que son iguales entre dos vectores, pero con m谩s peso en los elementos que est谩n presentes en ambos vectores.

- russellrao: esta m茅trica mide la cantidad de elementos que son iguales entre dos vectores dividida por el n煤mero total de elementos.

- seuclidean: esta m茅trica es similar a la distancia euclidiana, pero tiene en cuenta la varianza de cada dimensi贸n.

- sokalmichener: esta m茅trica mide la cantidad de elementos que son iguales entre dos vectores, pero con m谩s peso en los elementos que est谩n presentes en ambos vectores y menos peso en los elementos

# Visualizar matrices dispersas en Python

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.sparse import random

# Crear matriz dispersa aleatoria
n = 10
density = 0.3
matrix = random(n, n, density=density)
print(matrix.shape)
print("\n")
print(matrix)
```

```python
# Convertir matriz dispersa a una matriz densa
dense_matrix = matrix.todense()
print(dense_matrix)
```

```python
# Graficar matriz densa
plt.imshow(dense_matrix, cmap='gray', interpolation='nearest')
plt.show()
```

```python
# Graficar matriz dispersa
plt.spy(matrix, markersize=5)
plt.show()
```

matrix.todense() es un m茅todo de la clase scipy.sparse.csr_matrix que convierte una matriz dispersa en una matriz densa. En otras palabras, convierte una matriz representada de manera dispersa, en la que solo se almacenan los elementos distintos de cero, en una matriz en la que se almacenan todos los elementos.

El m茅todo todense() devuelve una matriz densa en formato de matriz de Numpy (clase numpy.matrix). Si la matriz dispersa original ten铆a forma (m, n), la matriz densa resultante tendr谩 forma (m, n) tambi茅n. Todos los elementos de la matriz densa, tanto los elementos distintos de cero como los ceros, se almacenan en la memoria.

Es importante tener en cuenta que la conversi贸n de una matriz dispersa a una matriz densa puede ser costosa en t茅rminos de uso de memoria y tiempo de ejecuci贸n, especialmente si la matriz dispersa es grande. Por lo tanto, en algunos casos, puede ser m谩s eficiente trabajar con la matriz dispersa en su forma original y utilizar m茅todos y funciones espec铆ficas para matrices dispersas.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.sparse import random

# Crear matriz dispersa aleatoria
n = 1000
density = 0.1
matrix = random(n, n, density=density)

# Convertir matriz dispersa a una matriz densa
dense_matrix = matrix.todense()

# Graficar matriz densa
plt.imshow(dense_matrix, cmap='gray', interpolation='nearest')
plt.show()

# Graficar matriz dispersa
plt.spy(matrix, markersize=.05)
plt.show()
```
