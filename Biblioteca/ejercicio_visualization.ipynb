## Librerías gráficas de Data Science

Las librerías gráficas de Data Science más utilizadas en Python son:

- Matplotlib: es una librería de trazado 2D que se utiliza para crear gráficos estáticos, animaciones y visualizaciones en línea.

- Seaborn: es una librería de visualización de datos basada en Matplotlib que proporciona una interfaz más fácil de usar para crear gráficos estadísticos.

- Plotly: es una librería de trazado interactivo que permite crear gráficos en 3D y visualizaciones en línea.

- Bokeh: es una librería de visualización interactiva para crear gráficos en línea y aplicaciones web.

- ggplot: es una librería de visualización de datos basada en R y portada a python que se basa en el paquete ggplot2 de R y proporciona una sintaxis similar para crear gráficos de alta calidad.


## Matplotlib

Veamos un ejemplo de cómo crear un gráfico de línea simple utilizando la librería Matplotlib en Python:

```python
import matplotlib.pyplot as plt

# datos
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 3, 10]

# crear el gráfico
plt.plot(x, y)

# agregar título y etiquetas de eje
plt.title('Gráfico de línea simple')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# mostrar el gráfico
plt.show()
```

Este código crea un gráfico de línea simple con los datos especificados en las variables x e y, agrega un título y etiquetas de eje al gráfico, y luego muestra el gráfico utilizando el método show(). El resultado sería una linea que se traza entre los puntos (1,2), (2,4), (3,6), (4,3) y (5,10)

## Seaborn

¿Cómo sería esto mismo con Seaborn?

```python
import seaborn as sns
import matplotlib.pyplot as plt # atención! seaborn necesita matplotlib

# crear el gráfico
sns.lineplot(x = x, y = y)

# agregar título y etiquetas de eje
plt.title('Gráfico de línea simple')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# mostrar el gráfico
plt.show()
```

En este caso, se utiliza el método lineplot() de Seaborn en lugar de plot() de Matplotlib para crear el gráfico de línea, y los mismos métodos de Matplotlib para agregar título y etiquetas de eje y mostrar el gráfico.

En general, Seaborn tiene una interfaz más amigable y se utiliza para crear gráficos estadísticos, como histogramas, gráficos de caja, mapas de calor, entre otros, y se integra muy bien con Pandas para trabajar con dataframes.

## Plotly

¿Y cómo sería este mismo ejemplo con Plotly?

Primero es necesario instalar las librerías:

```python
import plotly.express as px

# crear el gráfico
fig = px.line(x=x, y=y)

# agregar título y etiquetas de eje
fig.update_layout(title='Gráfico de línea simple',
                  xaxis_title='Eje X',
                  yaxis_title='Eje Y')

# mostrar el gráfico
fig.show()
```

En este caso, se utiliza el método px.line() de Plotly para crear el gráfico de línea, y se utiliza el método update_layout() para agregar título y etiquetas de eje al gráfico, y luego se utiliza el método show() para mostrar el gráfico. El resultado sería similar a los anteriores con una funcionalidad interactiva como hover, zoom, etc.

Hover es una funcionalidad en las visualizaciones interactivas que permite mostrar información adicional sobre un punto o elemento específico en un gráfico al mover el cursor sobre él. Esta información adicional puede incluir valores de datos, etiquetas, descripciones, etc. Es una forma de proporcionar contexto adicional al usuario y ayudar a entender mejor los datos que se están representando.

En las librerías de visualización como Plotly, Bokeh, se pueden agregar etiquetas o tooltips en los puntos o elementos para que al hacer hover se muestre información adicional sobre ese punto o elemento en particular.

En general, Plotly es una librería de visualización interactiva que permite crear gráficos en 3D y visualizaciones en línea, y es muy útil para crear visualizaciones dinámicas y de alto impacto.

## Bokeh

¿Cómo sería este ejemplo con Bokeh?

```python
from bokeh.plotting import figure, show

# crear el gráfico
p = figure(title='Gráfico de línea simple')
p.line(x, y)
p.xaxis.axis_label = 'Eje X'
p.yaxis.axis_label = 'Eje Y'

# mostrar el gráfico
show(p)
```

En este caso, se utiliza el método figure() de Bokeh para crear un objeto de gráfico vacío, y se utiliza el método line() para dibujar una línea en el gráfico con los datos especificados en las variables x e y, se utiliza xaxis.axis_label y yaxis.axis_label para agregar etiquetas al eje x e y y el método show() para mostrar el gráfico. El resultado permite interactuar con el gráfico en una aplicación web con herramientas como hover, zoom, etc.

En general, Bokeh es una librería de visualización interactiva para crear gráficos en línea y aplicaciones web, es muy útil para crear visualizaciones dinámicas y de alto impacto, y se integra muy bien con otras librerías como Pandas, NumPy, entre otras.

## ggplot

Y por último, ¿cómo sería este ejemplo con ggplot?

Requiere instalar previamente la librería:

pip install ggplot

```python
from ggplot import *
import pandas as pd

# datos
df = pd.DataFrame({'x':x, 'y':y})

# crear el gráfico
p = ggplot(df, aes(x='x', y='y')) + geom_line() + ggtitle("Gráfico de línea simple") + xlab("Eje X") + ylab("Eje Y")

print(p)
```

En este caso se utiliza la librería ggplot para crear el gráfico, la cual es una librería de visualización de datos basada en R y portada a python, y se basa en el paquete ggplot2 de R y proporciona una sintaxis similar para crear gráficos de alta calidad. Se utiliza un dataframe para almacenar los datos, con las columnas x e y, luego se utiliza el método ggplot() para crear un objeto de gráfico vacío, y se utiliza el método geom_line() para dibujar una línea en el gráfico con los datos especificados en las variables x e y, se utiliza ggtitle(), xlab() y ylab() para agregar título y etiquetas al eje x e y. El resultado debería ser una gráfica con una estética similar a la de ggplot2 de R.

## Histogramas

Un histograma es un gráfico que muestra la distribución de frecuencia de un conjunto de datos. Se utiliza para representar datos numéricos mediante columnas o barras, donde el eje horizontal representa las categorías de datos y el eje vertical representa la frecuencia de esos datos en esas categorías.

Vamos a trabajar con un dataset de salarios de científicos de datos, que encontramos en https://www.kaggle.com/datasets/saurabhshahane/data-science-jobs-salaries

```python
import pandas as pd

# IMPORTAMOS NUESTRO ARCHIVO CSV Y LO CONVERTIMOS EN UN DATAFRAME
filename = "Data Science Jobs Salaries.csv"
df = pd.read_csv(filename, delimiter = ",")
df
```

```python
# Reducimos los datos a la columna "salary"
df = df[["salary"]]
df
```

```python
# Visualizamos las líneas generales del dataframe
print(df.describe())
```

```python
# Lo visualizamos como un histograma
plt.hist(df.salary, bins=100, density=True)
```

### Histograma en Seaborn

Para dibujar un histograma con seaborn, primero debe importar la biblioteca y establecer un estilo de gráfico. Luego, puede utilizar la función seaborn.histplot() para dibujar el histograma. El parámetro data debe especificar el conjunto de datos que se utilizará para el histograma, mientras que el parámetro kde se puede establecer en True si desea mostrar una curva de densidad sobre el histograma.

Nota: Una curva de densidad es una línea suave que se superpone en un histograma para indicar la distribución continua subyacente de los datos. Es una representación gráfica de la función de densidad de probabilidad de los datos. La función de densidad de probabilidad es una función matemática que describe cómo se distribuyen los datos dentro de un conjunto.

La curva de densidad se utiliza para dar una idea de la distribución de los datos, especialmente cuando los datos son continuos.

En el caso de un histograma, la curva de densidad se superpone en las barras del histograma, y ayuda a dar una idea de cómo se distribuyen los datos dentro de cada categoría. La altura de la curva en cualquier punto indica la probabilidad de que un dato caiga en esa región. La curva de densidad se ajusta automáticamente a los datos y su forma general puede indicar si los datos siguen una distribución normal u otra forma de distribución.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Establecer estilo de gráfico
sns.set_style("darkgrid")

# Dibujar histograma
sns.histplot(data=df, kde=True)

# Mostrar gráfico
plt.show()
```

### Dibujando un histograma con Plotly

Para dibujar un histograma con Plotly, primero hay que crear una figura vacía. Luego, se utiliza el método add_trace() para agregar un trazo de histograma a la figura. El argumento x del método add_trace() debe contener los datos para el eje x, y el argumento name se puede utilizar para especificar un título para el trazo.

```python
import plotly.graph_objs as go

# crear figura vacía
fig = go.Figure()

# agregar trazo de histograma
fig.add_trace(go.Histogram(x=df["salary"], name='Histograma'))

# mostrar figura
fig.show()
```

Fijarse que con Plotly, el dato x pasado a la función es sólo una columna de datos, no todo el dataframe.

### Histograma con Bokeh

Bokeh no dispone de funcionalidad propia para creación de histogramas, por lo que realizamos el cálculo previo de la longitud de las barras mediante una función matemática de numPy.

```python
import numpy as np
from bokeh.io import show, output_file
from bokeh.plotting import figure

hist, edges = np.histogram(df.salary, density=True, bins=50)

p = figure()
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], line_color="white")

output_file("hist.html")
show(p)
```

Este código primero importa las bibliotecas necesarias: numpy, bokeh.io y bokeh.plotting.

Utilizando el método np.histogram() de numpy se genera un histograma a partir de los datos de la columna salary del dataframe. El argumento density especifica que se desea calcular la densidad de probabilidad en lugar de la frecuencia absoluta. El argumento bins especifica el número de divisiones o barras que se desea en el histograma.

Se crea una figura vacía utilizando figure() de bokeh.plotting.

El método quad() se utiliza para agregar un trazo de histograma a la figura. El argumento top especifica los datos para el eje y, bottom especifica el valor mínimo para el eje y, left especifica los datos para el eje x en los bordes izquierdos de las barras y right especifica los datos para el eje x en los bordes derechos de las barras. El argumento line_color especifica el color de las líneas que delimitan las barras del histograma.

Utilizando el método output_file() se especifica el nombre del archivo donde se guardará el histograma, en este caso se guarda en un archivo llamado "hist.html".

Utilizando el método show() se muestra la figura en pantalla.

Este código genera un histograma en un archivo HTML, que se puede abrir con cualquier navegador web. El histograma se dibuja con barras de ancho uniforme y el eje x representa el rango de salarios, mientras que el eje y representa la densidad de probabilidad de los salarios.

### Histograma con ggplot

Para hacer un histograma con ggplot, primero se crea una figura vacía utilizando el método ggplot(). Luego, se usa el método geom_histogram() para agregar un trazo de histograma a la figura. El argumento aes(x=...) del método geom_histogram() debe contener el nombre de la columna de los datos a utilizar para el eje x.

```python
from ggplot import *

#crear histograma
p = ggplot(df, aes(x='salary')) + geom_histogram(binwidth=1)

# mostrar figura
print(p)
```
