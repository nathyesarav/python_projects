```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creamos una funcion.
def numerosDivisibles(n1, n2):
        resultado = []
        
        for n in range(n1, n2 + 1):
            if n % 7 == 0 and n % 5 != 0: # Ponemos signo de exclamacion para que nos de otro resultado diferente a 0.
                resultado.append(n)
        return resultado    
```

```python
# LLamamos la funcion pasandole los dos paramatres.
numeros = numerosDivisibles(2200, 3400)
print(numeros)
```

# EJERCICIO DE DICCIONARIO

```python
def diccionario(n):
    date = {}
    
    for i in range(1, n + 1):
        date[i] = i*i
    return date
```

```python
resultado = diccionario(25)
print(resultado)
```

# Ejercicio de knn

```python
# Cargar datos. Vamos a utilizar el conjunto de datos de flores Iris:
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)
# El parámetro header=None se utiliza para indicar que el conjunto de datos no tiene una fila de encabezado en el archivo CSV

# Pandas asigna nombres de columna predeterminados (0, 1, 2, etc.) a cada columna. Renombramos:
iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']

iris
```

```python
iris["class"].value_counts()
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
from sklearn.model_selection import train_test_split

X = iris.iloc[:, :-1] # todos los datos excepto la última columna
y = iris.iloc[:, -1] # sólo la última columna
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_test
```

```python
# Crear el modelo de knn
# La misma librería nos permite encontrar los puntos vecinos o más cercanos a uno dado
from sklearn.neighbors import NearestNeighbors
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(X_train, y_train)
```

```python
predLabel = knn.predict(X_test)
print(predLabel)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# inicializar modelo
knn = KNeighborsClassifier(n_neighbors=1)

# entrenar modelo
knn.fit(X_train, y_train)

# evaluar modelo
score = knn.score(X_train, y_train)
print("Precisión del modelo:", score)
```

```python
X_test
```

```python
predLabel = knn.predict(X_test)
predLabel
```

```python
# evaluar modelo
score = knn.score(X_test, y_test)
print("Precisión del modelo en test:", score)
```

```python
# Sacamos por pantalla los datos de test, junto con la predicción, los resultados esperados y booleano de acierto
df = X_test
df["prediccion"] = predLabel
df["real"] = y[X_test.index]
df["acierto"] = (df["prediccion"] == df["real"])

df
```

```python
setosa     = df[df["prediccion"] == "Iris-setosa"    ]
virginica  = df[df["prediccion"] == "Iris-virginica" ]
versicolor = df[df["prediccion"] == "Iris-versicolor"]

plt.scatter(setosa["sepal_width"],     setosa["sepal_length"],     color="red",    s=50)
plt.scatter(virginica["sepal_width"],  virginica["sepal_length"],  color="yellow", s=50)
plt.scatter(versicolor["sepal_width"], versicolor["sepal_length"], color="blue",   s=50)

plt.show()
```

```python
Rsetosa     = df[df["real"] == "Iris-setosa"    ]
Rvirginica  = df[df["real"] == "Iris-virginica" ]
Rversicolor = df[df["real"] == "Iris-versicolor"]

plt.scatter(setosa["sepal_width"],     setosa["sepal_length"],       color="red",    s=10, zorder=1)
plt.scatter(Rsetosa["sepal_width"],    Rsetosa["sepal_length"],      color="red",    s=50, zorder=0)

plt.scatter(virginica["sepal_width"],   virginica["sepal_length"],   color="yellow", s=10, zorder=1)
plt.scatter(Rvirginica["sepal_width"],  Rvirginica["sepal_length"],  color="yellow", s=50, zorder=0)

plt.scatter(versicolor["sepal_width"],  versicolor["sepal_length"],  color="blue",   s=10, zorder=1)
plt.scatter(Rversicolor["sepal_width"], Rversicolor["sepal_length"], color="blue",   s=50, zorder=0)

plt.show()
```

```python
floresMalClasificadas = df[df["acierto"] == False]

plt.scatter(setosa["sepal_width"],     setosa["sepal_length"],       color="red",    s=10, zorder=1)
plt.scatter(Rsetosa["sepal_width"],    Rsetosa["sepal_length"],      color="red",    s=50, zorder=0)

plt.scatter(virginica["sepal_width"],   virginica["sepal_length"],   color="yellow", s=10, zorder=1)
plt.scatter(Rvirginica["sepal_width"],  Rvirginica["sepal_length"],  color="yellow", s=50, zorder=0)

plt.scatter(versicolor["sepal_width"],  versicolor["sepal_length"],  color="blue",   s=10, zorder=1)
plt.scatter(Rversicolor["sepal_width"], Rversicolor["sepal_length"], color="blue",   s=50, zorder=0)

plt.scatter(floresMalClasificadas["sepal_width"], floresMalClasificadas["sepal_length"], color="green", s=500, zorder=2, marker='+')

plt.show()
```
