# Regresi칩n Cuadr치tica

### Generamos el dataset

```python
import numpy as np

# Definimos el intervalo de valores x
xs = np.linspace(-5, 5, 20)

# Generamos los valores y con una funci칩n cuadr치tica y algo de ruido
ys = xs**2 + np.random.uniform(-5, 5, len(xs))

print(xs, ys)
```

### Dibujamos el dataset

```python
import matplotlib.pyplot as plt

# Lo visualizamos
plt.plot(xs, ys)
plt.scatter(xs, ys, s=15)
```

### Ajuste polin칩mico de grado 2

```python
# Polynomial fit with degree = 2
model = np.poly1d(np.polyfit(xs, ys, 2))

# Add fitted polynomial line to scatterplot
polyline = np.linspace(-8, 8, 100)
plt.scatter(xs, ys)
plt.plot(polyline, model(polyline))
plt.show()
```

### Par치metros de la regresi칩n

```python
print(model)
```

### Extrapolaci칩n

Vamos a calcular la predicci칩n del modelo para x = 10

```python
x = 10
y = 1.029 * x**2 - 0.1886 * x - 1.32
y
```

```python
# Visualizamos la extrapolaci칩n
xs = np.append(xs, x)
ys = np.append(ys, y)

plt.scatter(xs, ys, s=15)

# Add fitted polynomial line to scatterplot
polyline = np.linspace(-8, 14, 100)
plt.plot(polyline, model(polyline))
plt.show()
```
