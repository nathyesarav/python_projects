# Estadística en Data Science

Algunos de los conceptos que debes conocer como data scientist son:

- Estadística descriptiva: es el estudio de las características numéricas de un conjunto de datos. Incluye medidas de tendencia central (como la **media**, la **mediana** y la **moda**) y medidas de dispersión (como la **varianza** y la **desviación estándar**).

- Distribuciones de probabilidad: son modelos matemáticos que se utilizan para describir la forma en que se distribuyen los datos. Algunas de las distribuciones más comunes incluyen la distribución normal, la distribución binomial y la distribución de Poisson.

- Inferencia estadística: es el proceso de sacar conclusiones sobre una población a partir de una muestra de datos. Incluye técnicas como la prueba de hipótesis y el intervalo de confianza.

- Regresión: es una técnica estadística que se utiliza para encontrar la relación entre dos variables. La regresión lineal es la forma más común de regresión y se utiliza para encontrar una línea que mejor se ajuste a los datos.

- Métodos de muestreo: son técnicas utilizadas para seleccionar una muestra de una población para su estudio. El muestreo aleatorio simple es la técnica más común. Ej.: df.sample(100)

- Análisis de series temporales: es una técnica estadística que se utiliza para analizar datos que varían con el tiempo. Incluye técnicas como el análisis de tendencias y el análisis de estacionalidad.

# Distribución binomial

Un **ensayo de Bernoulli** es un experimento aleatorio con dos posibles resultados, éxito o fracaso, en el que la probabilidad de éxito es constante en cada ensayo y se denota por "p". Este tipo de experimento aleatorio se utiliza comúnmente en estadística y teoría de probabilidades para modelar situaciones donde sólo hay dos posibles resultados.

Por ejemplo, **lanzar una moneda** es un ejemplo de un ensayo de Bernoulli, donde el éxito se define como obtener cara y el fracaso se define como obtener cruz. Otro ejemplo es el lanzamiento de un dado, donde el éxito se define como obtener un número par y el fracaso se define como obtener un número impar.

Los ensayos de Bernoulli se utilizan como base para la distribución binomial, que modela el número de éxitos en una serie de ensayos de Bernoulli independientes. La distribución **binomial** lleva ese nombre porque es una distribución que modela el número de éxitos en una serie de ensayos de Bernoulli independientes, donde cada ensayo tiene **sólo dos posibles resultados, éxito o fracaso**. El término "binomial" hace referencia a que hay dos posibilidades para cada ensayo, lo que implica que se trata de una distribución "bi" o "dual".

Vamos a generar una distribución binomial con 10 ensayos y una probabilidad de éxito del 0.5:

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt

n = 10
p = 0.5

# Generar los valores para la distribución binomial
x = np.arange(0, n+1)
y = binom.pmf(x, n, p)

# Crear la gráfica
plt.plot(x, y, 'bo', ms=8)
plt.vlines(x, 0, y, colors='b', lw=5)
plt.title('Distribución Binomial')
plt.xlabel('Número de éxitos')
plt.ylabel('Probabilidad')
plt.show()
```

La gráfica representa la probabilidad de obtener un número determinado de éxitos en una serie de 10 ensayos independientes con una probabilidad de éxito del 0.5 en cada ensayo.

En forma, es una distribución parecida a la distribución gaussiana. Sin embargo, la distribución binomial es discreta y se utiliza para modelar el número de éxitos en una serie de ensayos de Bernoulli, mientras que la distribución gaussiana es continua y se utiliza para modelar datos que se distribuyen como las medidas físicas, errores de medición, y fenómenos naturales, sociales y económicos como la estatura, el peso, las puntuaciones de prueba.

En la distribución gaussiana, la media y la desviación estándar son los parámetros clave que determinan la forma de la distribución. Los parámetros de la districuión binomial son la probabilidad de éxito y el número de ensayos.

Además, la distribución binomial puede tener una forma asimétrica si la probabilidad de éxito no es igual a la probabilidad de fracaso, mientras que la distribución gaussiana siempre tiene una forma simétrica alrededor de su media.

```python
n = 10
p = 0.2

# Generar los valores para la distribución binomial
x = np.arange(0, n+1)
y = binom.pmf(x, n, p)

# Crear la gráfica
plt.plot(x, y, 'bo', ms=8)
plt.vlines(x, 0, y, colors='b', lw=5)
plt.title('Distribución Binomial')
plt.xlabel('Número de éxitos')
plt.ylabel('Probabilidad')
plt.show()
```

# Distribución de Poisson

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro lambda
lam = 5

# Crear una distribución de Poisson con el parámetro lambda
poisson_dist = np.random.poisson(lam, 1000)

# Graficar la distribución de Poisson
plt.hist(poisson_dist, bins=20)
plt.xlabel('Número de eventos')
plt.ylabel('Frecuencia')
plt.title('Distribución de Poisson con lambda = {}'.format(lam))
plt.show()
```

La distribución de Poisson es una distribución **discreta** que se utiliza para modelar la ocurrencia de eventos raros y aleatorios en un intervalo de tiempo o espacio determinado. Debido a la naturaleza discreta de la distribución, los datos pueden no llenar uniformemente los bins del histograma y pueden aparecer espacios vacíos.

```python
# Parámetro lambda
lam = 5

# Crear una distribución de Poisson con el parámetro lambda
poisson_dist = np.random.poisson(lam, 1000)

# Graficar la distribución de Poisson
plt.hist(poisson_dist, bins=10)
plt.xlabel('Número de eventos')
plt.ylabel('Frecuencia')
plt.title('Distribución de Poisson con lambda = {}'.format(lam))
plt.show()
```

La distribución de Poisson se caracteriza por **un solo parámetro lambda**, que representa la tasa promedio de ocurrencia de eventos raros, mientras que la distribución normal se caracteriza por dos parámetros: la media (representada por la letra griega mu) y la desviación estándar (representada por la letra griega sigma). La media representa el valor central de la distribución, mientras que la desviación estándar mide la dispersión de los datos alrededor de la media.

Un ejemplo sencillo de caso donde se puede aplicar la distribución de Poisson es en el **conteo de accidentes de tráfico** en una intersección durante un periodo de tiempo determinado, como por ejemplo un mes.

Supongamos que queremos determinar la probabilidad de que ocurran cierto número de accidentes en una intersección durante un mes determinado. Podemos suponer que la tasa promedio de accidentes en la intersección es de 2 por mes, es decir, que esperamos que ocurran dos accidentes por mes en promedio.

Para modelar esta situación, podemos utilizar la distribución de Poisson con un parámetro lambda igual a 2. Entonces, la probabilidad de que ocurran k accidentes en un mes determinado se puede calcular utilizando la fórmula de la distribución de Poisson:

P(k accidentes) = (lambda^k * e^(-lambda)) / k!

Por ejemplo, la probabilidad de que ocurran exactamente 3 accidentes en un mes sería:

P(3 accidentes) = (2^3 * e^(-2)) / 3! = 0.1804

# Inferencia estadística

Obtenemos conclusiones sobre una población (muchos) a partir de una muestra (pocos) de datos. La prueba de hipótesis y el intervalo de confianza son las técnicas aplicables:



Ejemplos de situaciones en las que se pueden aplicar estas técnicas:

- Prueba de hipótesis: Supongamos que una empresa quiere determinar si el tiempo promedio de espera en la cola de atención al cliente en sus tiendas es mayor a 5 minutos. Para ello, se toma una muestra aleatoria de tiendas y se registra el tiempo de espera en cada una. Luego, se utiliza la prueba de hipótesis para determinar si hay suficiente evidencia estadística para rechazar la hipótesis nula (H0) de que el tiempo promedio de espera es igual o menor a 5 minutos en todas las tiendas. Si se rechaza H0, entonces se puede concluir que el tiempo promedio de espera es mayor a 5 minutos en la población de tiendas.

- Intervalo de confianza: Supongamos que una encuesta realizada a una muestra aleatoria de la población indica que el porcentaje de personas que apoya una cierta política es del 60%. Para determinar el intervalo de confianza para este resultado, se utiliza la técnica del intervalo de confianza. Por ejemplo, con un nivel de confianza del 95% se puede construir un intervalo de confianza del 55% al 65%. Esto significa que, con un nivel de confianza del 95%, se puede afirmar que el porcentaje de apoyo en la población general se encuentra dentro de este intervalo.

- Prueba de hipótesis: Un equipo de investigadores quiere determinar si un nuevo tratamiento para una enfermedad es efectivo. Para ello, se selecciona una muestra aleatoria de pacientes y se divide en dos grupos: uno que recibe el nuevo tratamiento y otro que recibe un placebo. Luego, se utiliza la prueba de hipótesis para determinar si hay suficiente evidencia estadística para rechazar la hipótesis nula (H0) de que no hay diferencia en la tasa de curación entre los dos grupos. Si se rechaza H0, entonces se puede concluir que el nuevo tratamiento es efectivo.

- Intervalo de confianza: Un fabricante de juguetes quiere determinar el peso promedio de una nueva línea de juguetes para asegurarse de que cumple con los estándares de seguridad. Para ello, se toma una muestra aleatoria de juguetes y se mide su peso. Luego, se utiliza la técnica del intervalo de confianza para determinar el intervalo de pesos en el que se encuentra el peso promedio de la población de juguetes con un cierto nivel de confianza. Por ejemplo, con un nivel de confianza del 99%, se puede construir un intervalo de confianza del 10 gramos al 20 gramos. Esto significa que, con un nivel de confianza del 99%, se puede afirmar que el peso promedio de la población de juguetes se encuentra dentro de este intervalo.
