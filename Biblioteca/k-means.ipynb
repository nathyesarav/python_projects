# K-means (clustering no supervisado)

En el aprendizaje automtico no supervisado, la mquina aprende a partir de datos sin la ayuda de etiquetas o respuestas conocidas previamente. En otras palabras, la mquina no tiene una salida deseada para cada entrada de datos que recibe. En su lugar, la mquina busca patrones y estructuras en los datos por s misma.

El objetivo del aprendizaje no supervisado es encontrar estructuras y patrones interesantes en los datos que puedan ser tiles para el anlisis o la toma de decisiones. Algunas tcnicas comunes de aprendizaje no supervisado incluyen el clustering (agrupamiento) y la reduccin de dimensionalidad. En el clustering, la mquina agrupa automticamente los datos similares en grupos separados. En la reduccin de dimensionalidad, la mquina busca formas de reducir la cantidad de variables o caractersticas en los datos sin perder demasiada informacin valiosa.

```python
import pandas as pd
```

```python
# Cargar datos. Vamos a utilizar el conjunto de datos de flores Iris:
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)
# El parmetro header=None se utiliza para indicar que el conjunto de datos no tiene una fila de encabezado en el archivo CSV
# Pandas asigna nombres de columna predeterminados (0, 1, 2, etc.) a cada columna. Renombramos:
iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris
```

We will refer to the rows of the matrix as samples, and the number of rows as n_samples and the the columns of the matrix as features, and the number of columns as n_features.

# Features matrix (X)

This table layout makes clear that the information can be thought of as a two-dimensional numerical array or matrix, called the features matrix with shape [n_samples, n_features]

```python
X = iris.iloc[:,:4] # no usamos la columna de clase, porque es una informacin que no queremos darle al modelo
X
```

# Target array (y)

In addition to the feature matrix X, we also generally work with a label or target array, which by convention we will usually call y. The target array is usually one dimensional, with length n_samples, and is generally contained in a NumPy array or Pandas Series.

```python
# Target names
y = iris["class"]
y.value_counts()
```

En la regresin lineal, el objetivo es predecir un valor numrico continuo (la variable dependiente) en funcin de una o ms variables predictoras (las variables independientes). El modelo se entrena utilizando un conjunto de datos etiquetados, donde la variable dependiente est presente en los datos de entrenamiento. # Machine Learning Terminology

- Each row is an observation (also known as : sample, example, instance, record)

- Each column is a feature (also known as: Predictor, attribute, Independent Variable, input, regressor, Covariate)

Las caractersticas fsicas de las flores tienden a ser similares entre sus especies. Por tanto, el objetivo de k-means es detectar los tipos de flor (clase) por similitud con otras instancias de la misma especie.

No dividimos los datos en test y train, porque no hay entrenamiento ni aprendizaje. Los modelos no supervisados son algoritmos que directamente ofrecen una prediccin y el paso de entrenamiento previo no va a requerir un subconjunto "train" sino todo el dataset completo. En esa fase de entrenamiento los modelos no supervisados buscan una convergencia de los parmetros internos del modelo, en el caso de k-means, esos parmetros internos son los centroides.

***

## Crear el modelo de k-means

```python
# Anlisis preliminar para detectar visualmente cuntos clusters podra haber
import plotly.express as px
fig = px.scatter_3d(iris, x='sepal_length', y='sepal_width', z='petal_width')
fig.update_layout(height=800)  # ajusta la altura del grfico 3d
fig.show()
```

```python
from sklearn.cluster import KMeans
# Creamos una instancia de la clase KMeans y especificamos el nmero de clsteres (k)
# que deseamos crear. Visualmente hemos detectado 2 clsteres:
kmeans = KMeans(n_clusters=2)
```

Ajustamos el modelo K-means a los datos utilizando el mtodo fit() y obtenemos las etiquetas de clster asignadas a cada muestra utilizando el mtodo predict():

```python
kmeans.fit(X) # el modelo busca los centroides mediante convergencia
y_kmeans = kmeans.predict(X) # el modelo determina el cluster correspondiente al dato segn su distancia a los centroides
```

```python
y_kmeans
```

```python
# Visualizamos los resultados utilizando un grfico de dispersin.
# Por ejemplo, podemos graficar las caractersticas 0 y 1 del conjunto de datos
# y colorear cada punto segn su etiqueta de clster:
import matplotlib.pyplot as plt

plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y_kmeans, cmap='viridis')
plt.xlabel('Caracterstica 0')
plt.ylabel('Caracterstica 1')
plt.show()
```

```python
# Hay dos centroides que son vectores de 4 dimensiones porque hay 4 features
centroids = kmeans.cluster_centers_
centroids
```

```python
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y_kmeans, cmap='viridis')

plt.scatter(centroids[0, 0], centroids[0, 1], s=300, color="blue")
plt.scatter(centroids[1, 0], centroids[1, 1], s=300, color="green")

plt.xlabel('Caracterstica 0')
plt.ylabel('Caracterstica 1')
plt.show()
```

```python
X.columns
```

```python
# Visualizamos los clusters y centroides en 3D
fig = px.scatter_3d(X, x='sepal_length', y='sepal_width', z='petal_width', color=y_kmeans, symbol=y_kmeans, opacity=0.7)

import plotly.graph_objs as go
fig.add_trace(go.Scatter3d(x=centroids[:, 0], y=centroids[:, 1], z=centroids[:, 3],
                           name='centroides', mode='markers',
                           marker=dict(size=10, color='black', opacity=1)))

fig.update_layout(height=800)  # ajusta la altura del grfico 3d
fig.show()
```

```python
# Visualizamos en 3D las clases etiquetadas y los centroides de k-means con k=2
fig = px.scatter_3d(iris, x='sepal_length', y='sepal_width', z='petal_width', color=y, symbol=y, opacity=0.7)
import plotly.graph_objs as go
fig.add_trace(go.Scatter3d(x=centroids[:, 0], y=centroids[:, 1], z=centroids[:, 3],
                           name='centroides', mode='markers',
                           marker=dict(size=10, color='black', opacity=1)))

fig.update_layout(height=800)  # ajusta la altura del grfico 3d
fig.show()
```

```python
dateparse = lambda x: pd.datetime.strptime(x, '%m/%d/%Y %H:%M')
data = pd.read_csv('OnlineRetail.csv', parse_dates=['InvoiceDate'], date_parser=dateparse, encoding = 'unicode_escape')
```

```python
# userItem = df.pivot("CustomerID", "StockCode")["Quantity"]
userItem = data.pivot_table(index='CustomerID', columns='StockCode', values='Quantity', aggfunc='sum')
userItem
```

```python
# Reemplazamos NaNs con ceros (0): 
userItemZero = userItem.fillna(0)
```

```python
kmeans = KMeans(n_clusters=25)
kmeans.fit(userItemZero) # el modelo busca los centroides mediante convergencia
y_kmeans = kmeans.predict(userItemZero) # el modelo determina el cluster correspondiente al dato segn su distancia a los centroides
y_kmeans
```
