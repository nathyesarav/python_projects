# Violin Plot

En el workshop de hoy vamos a ver cmo utilizar grficos de tipo Violin Plot y lo aplicaremos en varios casos, entre ellos dos distribuciones: normal y exponencial.

***

El IQR (rango intercuartil) es una medida de la dispersin de los datos en un conjunto de datos. En el contexto de un violinplot, el IQR se refiere a la distancia entre el primer cuartil (Q1) y el tercer cuartil (Q3) de la distribucin de los datos.

El primer cuartil (Q1) es el valor que separa el 25% inferior de los datos del 75% superior, mientras que el tercer cuartil (Q3) es el valor que separa el 75% inferior del 25% superior de los datos. El rango intercuartil (IQR) se calcula como la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1).

En un violinplot, el ancho de la caja que se encuentra en el centro del grfico representa el IQR, es decir, la distancia entre el primer y tercer cuartil. La mediana se representa mediante una lnea en la caja.

El IQR es til porque proporciona informacin sobre la variabilidad de los datos. Si el IQR es pequeo, significa que los datos estn muy concentrados alrededor de la mediana, mientras que si el IQR es grande, significa que los datos estn muy dispersos.

# Distribucin exponencial




Con la librera NumPy de Python vamos a generar una muestra de 10.000 nmeros aleatorios de una distribucin exponencial con un parmetro de escala de 20.

La distribucin exponencial es una distribucin de probabilidad continua que se utiliza para modelar el tiempo de espera entre eventos. En esta distribucin, la tasa de ocurrencia de eventos es constante, pero el tiempo entre eventos es aleatorio y se distribuye exponencialmente.

El parmetro de escala, que en este caso es 20, determina el promedio de la distribucin exponencial. Es decir, el tiempo medio entre eventos es igual a este parmetro de escala.

````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1234)
data = np.random.exponential(scale=20, size=10000)
data
````

````python
plt.plot(data)
````

````python
plt.scatter(range(0,10000), data, s=1)
````

La distribucin exponencial se usa para describir el tiempo entre eventos que ocurren de manera aleatoria, como por ejemplo, el tiempo entre dos vehculos que pasan por una calle o el tiempo entre dos rayos en una tormenta.

Imagina que estamos viendo una carrera de coches en la televisin y queremos saber cunto tiempo pasar entre cada coche que cruza la lnea de meta. La distribucin exponencial nos dice que es muy probable que pase un tiempo corto entre dos coches (como unos pocos segundos), pero tambin es posible que pase un tiempo mucho ms largo (como unos pocos minutos o incluso horas).

El tiempo promedio entre dos eventos, que se llama "parmetro de escala", es lo que se usa para calcular la forma de la distribucin exponencial. Por ejemplo, si el parmetro de escala es 5 segundos, es ms probable que pase un tiempo corto entre dos eventos, pero todava es posible que pase un tiempo mucho ms largo.

Es importante saber que la distribucin exponencial es solo una forma de describir los tiempos entre eventos, y que no es aplicable a todos los casos. Pero es til cuando queremos hacer predicciones sobre el tiempo que pasar hasta que ocurra algo.

````python
plt.hist(data, bins=200)
````

````python
import seaborn as sns
sns.violinplot(data=data)
plt.show()
````

# Distribucin Gaussiana

Para crear una distribucin gaussiana utilizando la librera NumPy de Python, podemos utilizar la funcin np.random.normal() que nos permite generar nmeros aleatorios que siguen una distribucin normal (tambin conocida como Distribucin Gaussiana o Campana de Bell).

La distribucin normal es una distribucin de probabilidad que tiene una forma de campana y se utiliza para modelar muchos fenmenos en la naturaleza y en la ciencia. En la distribucin normal, la mayora de los valores se encuentran cerca de la media (o promedio) y disminuyen a medida que nos alejamos de ella.

La funcin np.random.normal() tiene tres parmetros principales:

- loc: la media de la distribucin.

- scale: la desviacin estndar de la distribucin.

- size: el tamao de la muestra que queremos generar.

````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1234)
mu, sigma = 0, 20 # media y desviacin estndar
data = np.random.normal(mu, sigma, 10000)

data
````

````python
plt.scatter(range(0,10000), data, s=1)

# 10 ticks inclinados 45 grados
ticks = np.linspace(0, 10000, 11)
plt.xticks(ticks, rotation=45)

plt.show()
````

````python
plt.hist(data, bins=200)
````

````python
import seaborn as sns

sns.violinplot(data=data)
plt.show()
````

La distribucin exponencial y la distribucin gaussiana (tambin conocida como distribucin normal) son dos distribuciones de probabilidad diferentes que se utilizan para modelar distintos tipos de fenmenos.

La principal diferencia entre la distribucin exponencial y la distribucin gaussiana es que la distribucin exponencial describe el tiempo entre eventos que ocurren de manera aleatoria, mientras que la distribucin gaussiana describe la distribucin de valores continuos, como pueden ser la altura de las personas o las calificaciones en un examen.

Otra diferencia importante entre ambas distribuciones es la forma de su curva de probabilidad. La distribucin exponencial tiene una forma de "cola larga", donde la probabilidad de que ocurra un evento disminuye rpidamente con el tiempo, pero an es posible que ocurra despus de un tiempo muy largo. En cambio, la distribucin gaussiana tiene una forma de campana, donde la mayora de los valores se encuentran cerca de la media y disminuyen a medida que nos alejamos de ella.

La distribucin exponencial se caracteriza por un solo parmetro llamado "tasa" o "ndice", que determina la frecuencia con la que ocurren los eventos. Por otro lado, la distribucin gaussiana se caracteriza por dos parmetros: la media, que es el valor central de la distribucin, y la desviacin estndar, que mide la dispersin de los valores alrededor de la media.

# Aplicacin de violin plot en ventas por pases

````python
from datetime import datetime, timedelta
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize': (10, 5), 'figure.dpi': 120})
import numpy as np
import seaborn as sns
````

````python
dateparse = lambda x: pd.datetime.strptime(x, '%m/%d/%Y %H:%M')
df = pd.read_csv('OnlineRetail.csv', parse_dates=['InvoiceDate'], date_parser=dateparse, encoding = 'unicode_escape')
````

````python
df.head()
````

````python
compras_por_pais = df['Country'].value_counts()
compras_por_pais
````

````python
# Limit by countries
# df2 = df[df["Country"] in ["United Kingdom", "France", "Spain"]]
# df2 = df[df["Country"].isin(["United Kingdom", "France", "Spain"])]
df2 = df[df["Country"].isin(compras_por_pais[:10].keys())] # Los 10 pases con ms ventas
df2
````

````python
# Limit by removing outliers
# df2 = df2[df2["Quantity"] < 100 & df2["Quantity"] > -100]
df2 = df2[(df2["Quantity"] < 30) & (df2["Quantity"] > -2)]
````

````python
# Dibujar violinplots
sns.violinplot(x="Country", y="Quantity", data=df2)
plt.xticks(rotation=45)
plt.show()
````

El **violinplot** es un tipo de grfico que muestra la distribucin de los datos y su densidad en un conjunto de valores. A continuacin, se presentan algunas conclusiones que puedes extraer de un grfico violinplot:

- Distribucin de los datos: el violinplot permite identificar la forma de la distribucin de los datos. Por ejemplo, si la distribucin es simtrica o sesgada hacia un lado, o si hay valores atpicos o extremos.

- Densidad de los datos: el ancho del violinplot representa la densidad de los datos en esa parte de la distribucin. Un rea ms ancha indica que hay ms datos en esa parte de la distribucin, mientras que un rea ms estrecha indica que hay menos datos.

- Comparacin de distribuciones: el violinplot tambin se puede utilizar para comparar la distribucin de los datos entre dos o ms grupos. Por ejemplo, si los violinplots de dos grupos se superponen, esto puede indicar que las distribuciones son similares, mientras que si estn separados, esto puede indicar que las distribuciones son diferentes.

- Tendencia central: aunque el violinplot no proporciona una medida de tendencia central directamente, se puede utilizar para inferir la presencia de una moda o punto de inflexin en la distribucin de los datos.

# Visualizacin de la relacin cantidad/precio

Grfico de dispersin (scatter plot): Este tipo de grfico muestra cada observacin como un punto en un sistema de coordenadas. Podemos representar la cantidad en el eje x y el precio en el eje y. De esta forma, podemos ver la relacin entre las dos variables y detectar posibles patrones o tendencias.

````python
plt.scatter(df2["UnitPrice"], df2["Quantity"], s=1)
````

````python
plt.scatter(df2["UnitPrice"], df2["Quantity"], s=1)
plt.xscale("log")
plt.show()
````

````python
# Dibujar miles de violinplots
sns.violinplot(x="UnitPrice", y="Quantity", data=df2)

# establecer los lmites del eje x e y
plt.xlim([0, 1000])
plt.ylim([-20, 40])

# 10 ticks inclinados 45 grados
ticks = np.linspace(0, 1000, 11)
plt.xticks(ticks, rotation=45)

plt.show()
````

Vamos a utilizar la funcin "kdeplot" de Seaborn para crear un **grfico de densidad**.

Los argumentos "df2['UnitPrice']" y "df2['Quantity']" especifican las variables a representar. El argumento "shade=True" indica que deseamos sombrear el rea bajo la curva de densidad. El argumento "cmap='Blues'" especifica la paleta de colores a utilizar. Tambin se definen etiquetas de los ejes y el ttulo del grfico.

La salida es un grfico de densidad que muestra la relacin entre la cantidad y el precio unitario. La densidad se muestra como una superficie continua de colores que varan segn el valor de la densidad en cada punto. Este tipo de grfico nos permite ver cmo se distribuyen los datos en la regin donde hay mayor densidad y dnde hay menor densidad.

````python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(df2["UnitPrice"], df2["Quantity"], shade=True, cmap="Blues")
plt.xlabel("Precio Unitario")
plt.ylabel("Cantidad")
plt.title("Grfico de Densidad de Cantidad y Precio Unitario")
plt.show()
````

Casi no se visualizan los datos:

- Ancho de banda demasiado grande: Si el ancho de banda utilizado para estimar la densidad es demasiado grande, la curva de densidad se suavizar demasiado y puede que no muestre bien la variabilidad en los datos. Se puede ajustar el ancho de banda (bw_method) para que sea ms pequeo y as obtener una curva de densidad ms precisa.

Para ampliar el ancho de banda (bandwidth) en un grfico de densidad generado con kdeplot, se puede ajustar el valor del parmetro bw_method. Este parmetro controla cmo se calcula el ancho de banda para estimar la densidad.

Por defecto, bw_method est establecido en "scott", que utiliza una regla heurstica para calcular el ancho de banda basado en la desviacin estndar de los datos. Si se quiere ampliar el ancho de banda, se puede establecer bw_method en "silverman", que utiliza una regla ms suave para calcular el ancho de banda.

Tambin se puede ajustar manualmente el ancho de banda estableciendo el valor de bw_method en un nmero decimal. Un valor mayor de bw_method dar como resultado un ancho de banda ms amplio y una curva de densidad ms suave. Ejemplo:

```python
sns.kdeplot(data, bw_method=0.5)
```

- Funcin kernel inadecuada: Si la funcin kernel utilizada para estimar la densidad no es adecuada para los datos que se estn graficando, la curva de densidad puede resultar poco informativa. En este caso, se puede intentar cambiar la funcin kernel (kernel) para obtener una curva de densidad ms precisa.

````python
sns.kdeplot(df2["UnitPrice"], df2["Quantity"], shade=True, cmap="Blues", bw_method="silverman")
plt.xlabel("Precio Unitario")
plt.ylabel("Cantidad")
plt.title("Grfico de Densidad de Cantidad y Precio Unitario (bw_method=silverman)")
# plt.xscale("log")
plt.show()
````

````python
sns.kdeplot(df2["UnitPrice"], df2["Quantity"], shade=True, cmap="Blues", bw_method=0.5)
plt.xlabel("Precio Unitario")
plt.ylabel("Cantidad")
plt.title("Grfico de Densidad de Cantidad y Precio Unitario (bw_method=0.5)")
# plt.xscale("log")
plt.show()
````

````python
sns.kdeplot(df2["UnitPrice"], df2["Quantity"], shade=True, cmap="Blues", bw_method=5)
plt.xlabel("Precio Unitario")
plt.ylabel("Cantidad")
plt.title("Grfico de Densidad de Cantidad y Precio Unitario (bw_method=5)")
# plt.xscale("log")
plt.show()
````

# Violin Plot asimtrico

````python
import seaborn as sns
import pandas as pd

# Cargamos el conjunto de datos de pinginos,
# que contiene informacin sobre el peso de 
# diferentes especies de pinginos, separados por sexo.
df = sns.load_dataset('penguins')
df = df.dropna()
df
````

````python
sns.violinplot(x="sex", y="body_mass_g", data=df)
````

````python
massMale = df[df["sex"] == "Male"]["body_mass_g"]
massMale
````

````python
massFemale = df[df["sex"] == "Female"]["body_mass_g"]
massFemale
````

````python
from statsmodels.graphics.boxplots import violinplot
fig, ax = plt.subplots()
violinplot([massMale], positions=[0], show_boxplot=False, side='left', ax=ax, plot_opts={'violin_fc':'C0'})
violinplot([massFemale], positions=[0], show_boxplot=False, side='right', ax=ax, plot_opts={'violin_fc':'C1'})
plt.show()
````
