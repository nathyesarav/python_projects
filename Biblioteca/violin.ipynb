# Violin Plot

En el workshop de hoy vamos a ver cómo utilizar gráficos de tipo Violin Plot y lo aplicaremos en varios casos, entre ellos dos distribuciones: normal y exponencial.

***

El IQR (rango intercuartil) es una medida de la dispersión de los datos en un conjunto de datos. En el contexto de un violinplot, el IQR se refiere a la distancia entre el primer cuartil (Q1) y el tercer cuartil (Q3) de la distribución de los datos.

El primer cuartil (Q1) es el valor que separa el 25% inferior de los datos del 75% superior, mientras que el tercer cuartil (Q3) es el valor que separa el 75% inferior del 25% superior de los datos. El rango intercuartil (IQR) se calcula como la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1).

En un violinplot, el ancho de la caja que se encuentra en el centro del gráfico representa el IQR, es decir, la distancia entre el primer y tercer cuartil. La mediana se representa mediante una línea en la caja.

El IQR es útil porque proporciona información sobre la variabilidad de los datos. Si el IQR es pequeño, significa que los datos están muy concentrados alrededor de la mediana, mientras que si el IQR es grande, significa que los datos están muy dispersos.

# Distribución exponencial




Con la librería NumPy de Python vamos a generar una muestra de 10.000 números aleatorios de una distribución exponencial con un parámetro de escala de 20.

La distribución exponencial es una distribución de probabilidad continua que se utiliza para modelar el tiempo de espera entre eventos. En esta distribución, la tasa de ocurrencia de eventos es constante, pero el tiempo entre eventos es aleatorio y se distribuye exponencialmente.

El parámetro de escala, que en este caso es 20, determina el promedio de la distribución exponencial. Es decir, el tiempo medio entre eventos es igual a este parámetro de escala.

````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1234)
data = np.random.exponential(scale=20, size=10000)
data
````

````python
plt.plot(data)
````

````python
plt.scatter(range(0,10000), data, s=1)
````

La distribución exponencial se usa para describir el tiempo entre eventos que ocurren de manera aleatoria, como por ejemplo, el tiempo entre dos vehículos que pasan por una calle o el tiempo entre dos rayos en una tormenta.

Imagina que estamos viendo una carrera de coches en la televisión y queremos saber cuánto tiempo pasará entre cada coche que cruza la línea de meta. La distribución exponencial nos dice que es muy probable que pase un tiempo corto entre dos coches (como unos pocos segundos), pero también es posible que pase un tiempo mucho más largo (como unos pocos minutos o incluso horas).

El tiempo promedio entre dos eventos, que se llama "parámetro de escala", es lo que se usa para calcular la forma de la distribución exponencial. Por ejemplo, si el parámetro de escala es 5 segundos, es más probable que pase un tiempo corto entre dos eventos, pero todavía es posible que pase un tiempo mucho más largo.

Es importante saber que la distribución exponencial es solo una forma de describir los tiempos entre eventos, y que no es aplicable a todos los casos. Pero es útil cuando queremos hacer predicciones sobre el tiempo que pasará hasta que ocurra algo.

````python
plt.hist(data, bins=200)
````

````python
import seaborn as sns
sns.violinplot(data=data)
plt.show()
````

# Distribución Gaussiana

Para crear una distribución gaussiana utilizando la librería NumPy de Python, podemos utilizar la función np.random.normal() que nos permite generar números aleatorios que siguen una distribución normal (también conocida como Distribución Gaussiana o Campana de Bell).

La distribución normal es una distribución de probabilidad que tiene una forma de campana y se utiliza para modelar muchos fenómenos en la naturaleza y en la ciencia. En la distribución normal, la mayoría de los valores se encuentran cerca de la media (o promedio) y disminuyen a medida que nos alejamos de ella.

La función np.random.normal() tiene tres parámetros principales:

- loc: la media de la distribución.

- scale: la desviación estándar de la distribución.

- size: el tamaño de la muestra que queremos generar.

````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1234)
mu, sigma = 0, 20 # media y desviación estándar
data = np.random.normal(mu, sigma, 10000)

data
````

````python
plt.scatter(range(0,10000), data, s=1)

# 10 ticks inclinados 45 grados
ticks = np.linspace(0, 10000, 11)
plt.xticks(ticks, rotation=45)

plt.show()
````

````python
plt.hist(data, bins=200)
````

````python
import seaborn as sns

sns.violinplot(data=data)
plt.show()
````

La distribución exponencial y la distribución gaussiana (también conocida como distribución normal) son dos distribuciones de probabilidad diferentes que se utilizan para modelar distintos tipos de fenómenos.

La principal diferencia entre la distribución exponencial y la distribución gaussiana es que la distribución exponencial describe el tiempo entre eventos que ocurren de manera aleatoria, mientras que la distribución gaussiana describe la distribución de valores continuos, como pueden ser la altura de las personas o las calificaciones en un examen.

Otra diferencia importante entre ambas distribuciones es la forma de su curva de probabilidad. La distribución exponencial tiene una forma de "cola larga", donde la probabilidad de que ocurra un evento disminuye rápidamente con el tiempo, pero aún es posible que ocurra después de un tiempo muy largo. En cambio, la distribución gaussiana tiene una forma de campana, donde la mayoría de los valores se encuentran cerca de la media y disminuyen a medida que nos alejamos de ella.

La distribución exponencial se caracteriza por un solo parámetro llamado "tasa" o "índice", que determina la frecuencia con la que ocurren los eventos. Por otro lado, la distribución gaussiana se caracteriza por dos parámetros: la media, que es el valor central de la distribución, y la desviación estándar, que mide la dispersión de los valores alrededor de la media.

# Aplicación de violin plot en ventas por países

````python
from datetime import datetime, timedelta
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize': (10, 5), 'figure.dpi': 120})
import numpy as np
import seaborn as sns
````

````python
dateparse = lambda x: pd.datetime.strptime(x, '%m/%d/%Y %H:%M')
df = pd.read_csv('OnlineRetail.csv', parse_dates=['InvoiceDate'], date_parser=dateparse, encoding = 'unicode_escape')
````

````python
df.head()
````

````python
compras_por_pais = df['Country'].value_counts()
compras_por_pais
````

````python
# Limit by countries
# df2 = df[df["Country"] in ["United Kingdom", "France", "Spain"]]
# df2 = df[df["Country"].isin(["United Kingdom", "France", "Spain"])]
df2 = df[df["Country"].isin(compras_por_pais[:10].keys())] # Los 10 países con más ventas
df2
````

````python
# Limit by removing outliers
# df2 = df2[df2["Quantity"] < 100 & df2["Quantity"] > -100]
df2 = df2[(df2["Quantity"] < 30) & (df2["Quantity"] > -2)]
````

````python
# Dibujar violinplots
sns.violinplot(x="Country", y="Quantity", data=df2)
plt.xticks(rotation=45)
plt.show()
````

El **violinplot** es un tipo de gráfico que muestra la distribución de los datos y su densidad en un conjunto de valores. A continuación, se presentan algunas conclusiones que puedes extraer de un gráfico violinplot:

- Distribución de los datos: el violinplot permite identificar la forma de la distribución de los datos. Por ejemplo, si la distribución es simétrica o sesgada hacia un lado, o si hay valores atípicos o extremos.

- Densidad de los datos: el ancho del violinplot representa la densidad de los datos en esa parte de la distribución. Un área más ancha indica que hay más datos en esa parte de la distribución, mientras que un área más estrecha indica que hay menos datos.

- Comparación de distribuciones: el violinplot también se puede utilizar para comparar la distribución de los datos entre dos o más grupos. Por ejemplo, si los violinplots de dos grupos se superponen, esto puede indicar que las distribuciones son similares, mientras que si están separados, esto puede indicar que las distribuciones son diferentes.

- Tendencia central: aunque el violinplot no proporciona una medida de tendencia central directamente, se puede utilizar para inferir la presencia de una moda o punto de inflexión en la distribución de los datos.

# Visualización de la relación cantidad/precio

Gráfico de dispersión (scatter plot): Este tipo de gráfico muestra cada observación como un punto en un sistema de coordenadas. Podemos representar la cantidad en el eje x y el precio en el eje y. De esta forma, podemos ver la relación entre las dos variables y detectar posibles patrones o tendencias.

````python
plt.scatter(df2["UnitPrice"], df2["Quantity"], s=1)
````

````python
plt.scatter(df2["UnitPrice"], df2["Quantity"], s=1)
plt.xscale("log")
plt.show()
````

````python
# Dibujar miles de violinplots
sns.violinplot(x="UnitPrice", y="Quantity", data=df2)

# establecer los límites del eje x e y
plt.xlim([0, 1000])
plt.ylim([-20, 40])

# 10 ticks inclinados 45 grados
ticks = np.linspace(0, 1000, 11)
plt.xticks(ticks, rotation=45)

plt.show()
````

Vamos a utilizar la función "kdeplot" de Seaborn para crear un **gráfico de densidad**.

Los argumentos "df2['UnitPrice']" y "df2['Quantity']" especifican las variables a representar. El argumento "shade=True" indica que deseamos sombrear el área bajo la curva de densidad. El argumento "cmap='Blues'" especifica la paleta de colores a utilizar. También se definen etiquetas de los ejes y el título del gráfico.

La salida es un gráfico de densidad que muestra la relación entre la cantidad y el precio unitario. La densidad se muestra como una superficie continua de colores que varían según el valor de la densidad en cada punto. Este tipo de gráfico nos permite ver cómo se distribuyen los datos en la región donde hay mayor densidad y dónde hay menor densidad.

````python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(df2["UnitPrice"], df2["Quantity"], shade=True, cmap="Blues")
plt.xlabel("Precio Unitario")
plt.ylabel("Cantidad")
plt.title("Gráfico de Densidad de Cantidad y Precio Unitario")
plt.show()
````

Casi no se visualizan los datos:

- Ancho de banda demasiado grande: Si el ancho de banda utilizado para estimar la densidad es demasiado grande, la curva de densidad se suavizará demasiado y puede que no muestre bien la variabilidad en los datos. Se puede ajustar el ancho de banda (bw_method) para que sea más pequeño y así obtener una curva de densidad más precisa.

Para ampliar el ancho de banda (bandwidth) en un gráfico de densidad generado con kdeplot, se puede ajustar el valor del parámetro bw_method. Este parámetro controla cómo se calcula el ancho de banda para estimar la densidad.

Por defecto, bw_method está establecido en "scott", que utiliza una regla heurística para calcular el ancho de banda basado en la desviación estándar de los datos. Si se quiere ampliar el ancho de banda, se puede establecer bw_method en "silverman", que utiliza una regla más suave para calcular el ancho de banda.

También se puede ajustar manualmente el ancho de banda estableciendo el valor de bw_method en un número decimal. Un valor mayor de bw_method dará como resultado un ancho de banda más amplio y una curva de densidad más suave. Ejemplo:

```python
sns.kdeplot(data, bw_method=0.5)
```

- Función kernel inadecuada: Si la función kernel utilizada para estimar la densidad no es adecuada para los datos que se están graficando, la curva de densidad puede resultar poco informativa. En este caso, se puede intentar cambiar la función kernel (kernel) para obtener una curva de densidad más precisa.

````python
sns.kdeplot(df2["UnitPrice"], df2["Quantity"], shade=True, cmap="Blues", bw_method="silverman")
plt.xlabel("Precio Unitario")
plt.ylabel("Cantidad")
plt.title("Gráfico de Densidad de Cantidad y Precio Unitario (bw_method=silverman)")
# plt.xscale("log")
plt.show()
````

````python
sns.kdeplot(df2["UnitPrice"], df2["Quantity"], shade=True, cmap="Blues", bw_method=0.5)
plt.xlabel("Precio Unitario")
plt.ylabel("Cantidad")
plt.title("Gráfico de Densidad de Cantidad y Precio Unitario (bw_method=0.5)")
# plt.xscale("log")
plt.show()
````

````python
sns.kdeplot(df2["UnitPrice"], df2["Quantity"], shade=True, cmap="Blues", bw_method=5)
plt.xlabel("Precio Unitario")
plt.ylabel("Cantidad")
plt.title("Gráfico de Densidad de Cantidad y Precio Unitario (bw_method=5)")
# plt.xscale("log")
plt.show()
````

# Violin Plot asimétrico

````python
import seaborn as sns
import pandas as pd

# Cargamos el conjunto de datos de pingüinos,
# que contiene información sobre el peso de 
# diferentes especies de pingüinos, separados por sexo.
df = sns.load_dataset('penguins')
df = df.dropna()
df
````

````python
sns.violinplot(x="sex", y="body_mass_g", data=df)
````

````python
massMale = df[df["sex"] == "Male"]["body_mass_g"]
massMale
````

````python
massFemale = df[df["sex"] == "Female"]["body_mass_g"]
massFemale
````

````python
from statsmodels.graphics.boxplots import violinplot
fig, ax = plt.subplots()
violinplot([massMale], positions=[0], show_boxplot=False, side='left', ax=ax, plot_opts={'violin_fc':'C0'})
violinplot([massFemale], positions=[0], show_boxplot=False, side='right', ax=ax, plot_opts={'violin_fc':'C1'})
plt.show()
````
