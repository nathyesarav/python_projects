```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
grados = []
seno = []
coseno = []
tangente = []

for i in range(0,540):
	grados.append(i)
	radianes = (i* math.pi)/180
	seno.append(math.sin(radianes))
	coseno.append(math.cos(radianes))
	tangente.append(math.tan(radianes))

plt.ylim(-2,2)
plt.plot(grados,seno,"r")
plt.plot(grados,coseno,"b")
plt.plot(grados,tangente)
plt.show()
```

### Ruido sinusoide

```python
#Generando una seal:
tiempo = np.linspace(0, 1, 100) 
frecuencia = 3
Amplitud = 1 
signal = Amplitud * np.sin(2 * np.pi * frecuencia * tiempo)

#Generando ruido con 100 valores aleatorios de una distribucin normal con media 0 y desviacin estandar 0.2
noise = 0.2 * np.random.randn(100)  

# Agregando ruido a la seal
noisy_signal = signal + noise

# Graficando la seal original y la seal con ruido
plt.plot(tiempo, signal, label='Seal')
plt.plot(tiempo, noisy_signal, label='Seal con ruido')
plt.legend()
plt.show()
```
