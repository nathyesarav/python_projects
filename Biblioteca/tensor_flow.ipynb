# TensorFlow

### Entrenar modelos en Machine Learning

Entrenar un modelo en TensorFlow es necesario para que el modelo pueda aprender (Machine Learning) a partir de los datos y ser capaz de realizar predicciones o tomar decisiones en nuevos conjuntos de datos.

### Pipeline de ML

![](machine-learning-pipeline2.png)

En el pipeline/workflow de Aprendizaje Automático (ML), el entrenamiento del modelo es un paso esencial después de la recolección y limpieza de los datos, y antes de evaluar y desplegar el modelo en un sistema de producción.

**Pipeline** se traduce al español como tubería o canalización. **Workflow** se traduce como flujo de trabajo. Ambos significan la estructura o arquitectura del procesamiento de datos/input que realiza ML para obtener sus resultados/output **inferidos**.

La **inferencia** en aprendizaje automático (ML, por sus siglas en inglés) es el proceso de utilizar un modelo entrenado para hacer predicciones o tomar decisiones en nuevos conjuntos de datos. En otras palabras, es el proceso de aplicar el conocimiento aprendido por el modelo durante el entrenamiento para resolver problemas en un conjunto de datos desconocido. Por ejemplo, si un modelo fue entrenado para clasificar imágenes de perros y gatos, la inferencia sería el proceso de utilizar ese modelo para clasificar nuevas imágenes de animales.

### Usando TensorFlow para crear y entrenar un modelo de regresión lineal simple

```python
# Crea un modelo de regresión lineal simple que se ajusta a un conjunto de datos
# de entrenamiento y luego realiza una predicción para un solo punto de datos.
# El modelo se entrena utilizando el algoritmo Adam y la función de pérdida
# "mean squared error", y se ajusta durante 500 épocas.
import tensorflow as tf
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' # 3 = INFO, WARNING, and ERROR messages are not printed

# Creamos el modelo de regresión lineal
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=1, input_shape=[1]))

# Compilamos el modelo con una tasa de aprendizaje y una función de pérdida
model.compile(optimizer=tf.keras.optimizers.Adam(1), loss='mean_squared_error')

# Proporcionamos algunos datos de entrenamiento para el modelo
xs = [0, 1, 2, 3, 4]
ys = [1.1, 3.1, 5.2, 6.7, 8.9]

# Visualizamos los datos
from matplotlib import pyplot as plt

plt.plot(xs, ys)
plt.scatter(xs, ys, s=15)
```

```python
# Entrenamos el modelo sobre el conjunto de datos de entrenamiento
model.fit(xs, ys, epochs=500)

# Hacemos una predicción usando el modelo entrenado
print(model.predict([10]))
```

## tf.keras.layers.Dense

tf.keras.layers.Dense es una capa de red neuronal completamente conectada en TensorFlow. Esto significa que todos los nodos en la capa están conectados a todos los nodos en la capa siguiente.

Para crear una capa densa, se proporciona el número de unidades o nodos de la capa y la forma de la entrada. Por ejemplo, en el ejemplo anterior, se crea una capa densa con una sola unidad y una forma de entrada de una dimensión:

tf.keras.layers.Dense(units=1, input_shape=[1])

Esta capa tomará un tensor de una dimensión como entrada y lo mapeará a otro tensor de una dimensión como salida a través de una combinación lineal de pesos y sesgos.

La capa densa se utiliza comúnmente como una capa de salida en un modelo de redes neuronales, pero también se puede utilizar como una capa oculta.

Al ser ésta una red neuronal de sólo una neurona, es la más sencilla que podemos crear y tiene capacidad para modelar comportamientos lineales (regresión lineal).

#### ¿Qué son los pesos (weight) y el sesgo (bias) en una red neuronal?

Los pesos y el sesgo juegan un papel clave en la creación de una red neuronal. Los pesos se refieren a los valores asignados a las conexiones que se establecen entre los nodos de una red neuronal. Estos valores controlan la dirección y la cantidad de información que pasa de un nodo a otro. El sesgo, por otro lado, se refiere a la activación de una neurona. Esto significa que, incluso con una entrada nula, la neurona se puede activar. Es decir, el sesgo controla la sensibilidad de la neurona a la entrada. En conjunto, los pesos y el sesgo controlan la salida de una neurona, lo que permite a la red neuronal procesar datos y producir resultados.

**Nota: no confundir el sesgo de una neurona (umbral de activación) con el sesgo de una IA, que se refiere a la falta de imparcialidad en las fuentes de datos que alimentan el sistema.**

## ¿Qué es Keras?

Keras es una librería de redes neuronales de alto nivel escrita en Python. Fue diseñada para permitir una rápida experimentación con redes neuronales. Permite a los usuarios crear y entrenar modelos de redes neuronales de manera rápida y sencilla.

Keras se basa en TensorFlow, Theano o CNTK, lo que significa que puede ejecutarse en el backend de cualquiera de estas librerías. Esto permite a los usuarios aprovechar la potencia y escalabilidad de estas librerías mientras trabajan con una interfaz de usuario más amigable.

Una de las características más destacadas de Keras es que permite a los usuarios construir modelos de redes neuronales mediante la composición de capas, lo que permite experimentar con diferentes arquitecturas de redes neuronales. Además, Keras proporciona muchas funciones de pérdida, optimizadores y métricas que se pueden utilizar para entrenar y evaluar modelos de redes neuronales.

```python
# Creamos un modelo de regresión lineal innecesariamente más complejo, con más capas
# Definimos el modelo
model = tf.keras.Sequential()

# Añadimos la primera capa con 8 unidades y una forma de entrada de 1 dimensión
model.add(tf.keras.layers.Dense(units=8, input_shape=[1]))

# Añadimos la segunda capa con 4 unidades
model.add(tf.keras.layers.Dense(units=4))

# Compilamos el modelo con una tasa de aprendizaje y una función de pérdida
model.compile(optimizer=tf.keras.optimizers.Adam(1), loss='mean_squared_error')

# Proporcionamos algunos datos de entrenamiento para el modelo
xs = [0, 1, 2, 3, 4]
ys = [1.1, 3.1, 5.2, 6.7, 8.9]

# Entrenamos el modelo sobre el conjunto de datos de entrenamiento
model.fit(xs, ys, epochs=500)

# Hacemos una predicción usando el modelo entrenado
print(model.predict([10]))
```

## Red neuronal secuencial

tf.keras.Sequential es una clase de Keras que se utiliza para crear modelos de redes neuronales secuenciales. Un modelo de redes neuronales secuencial es un tipo de modelo que consiste en una secuencia de capas de redes neuronales.

Además de los modelos secuenciales, Keras también proporciona la opción de crear modelos de redes neuronales basados en la clase Model de Keras, que es una clase más general que permite un mayor control sobre la arquitectura del modelo.

Para crear un modelo basado en Model, primero debes definir las capas de tu modelo y luego utilizar estas capas para crear el modelo. Por ejemplo:

```python
# Definimos las capas de nuestro modelo
inputs = tf.keras.Input(shape=(3,))
x = tf.keras.layers.Dense(8, activation='relu')(inputs)
outputs = tf.keras.layers.Dense(4, activation='softmax')(x)

# Creamos el modelo a partir de las capas
model = tf.keras.Model(inputs=inputs, outputs=outputs)
```

En este ejemplo, hemos creado un modelo con dos capas densas. La primera capa tiene 8 unidades y una función de activación de ReLU, mientras que la segunda capa tiene 4 unidades y una función de activación de softmax.

Una vez que hayas creado tu modelo, puedes compilarlo y entrenarlo utilizando el método compile y el método fit, respectivamente. También puedes evaluar el rendimiento del modelo utilizando el método evaluate.

Además de los modelos secuenciales y basados en Model, Keras también proporciona otras opciones para crear modelos, como los modelos funcionales y los modelos basados en el API de alto nivel.

## ¿Qué otras funciones de pérdida hay además de mean_squared_error?

En Keras, hay una amplia variedad de funciones de pérdida disponibles para elegir, que se pueden utilizar para medir el error del modelo durante el entrenamiento. Algunas de las funciones de pérdida más comunes incluyen:

- mean_squared_error: calcula la pérdida media del cuadrado del error entre las predicciones del modelo y los valores objetivo.

- mean_absolute_error: calcula la pérdida media del valor absoluto del error entre las predicciones del modelo y los valores objetivo.

- categorical_crossentropy: utilizada para problemas de clasificación multiclase. Calcula la pérdida cruzada entre las predicciones del modelo y los valores objetivo.

- binary_crossentropy: utilizada para problemas de clasificación binaria. Calcula la pérdida cruzada entre las predicciones del modelo y los valores objetivo.

- sparse_categorical_crossentropy: similar a categorical_crossentropy, pero utilizada cuando los valores objetivo son enteros en lugar de vectores one-hot.

Éstas son algunas de las funciones de pérdida disponibles en Keras. Puedes encontrar una lista completa de las funciones de pérdida disponibles en la documentación de Keras: https://www.tensorflow.org/api_docs/python/tf/keras/losses

```python
import numpy as np

# Definimos el intervalo de valores x
xs = np.linspace(-5, 5, 20)

# Generamos los valores y con una función cuadrática y algo de ruido
ys = xs**2 + np.random.uniform(-5, 5, len(xs))

print(xs, ys)
```

```python
# Lo visualizamos
plt.plot(xs, ys)
plt.scatter(xs, ys, s=15)
```

```python
# Red neuronal capaz de extrapolar la secuencia parabólica de
# xs e ys definida anteriormente

# Definimos el modelo secuencial
model = tf.keras.Sequential()

# Añadimos una capa densa con 8 unidades y una forma de entrada de 1 dimensión
model.add(tf.keras.layers.Dense(units=8, input_shape=[1]))

# Añadimos otra capa densa con 4 unidades
model.add(tf.keras.layers.Dense(units=4))

# Añadimos una capa de salida con una sola unidad
model.add(tf.keras.layers.Dense(units=1))

# Compilamos el modelo con una tasa de aprendizaje y una función de pérdida
model.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss='mean_squared_error')

# Entrenamos el modelo sobre el conjunto de datos de entrenamiento
model.fit(xs, ys, epochs=500)
```

En este ejemplo, hemos creado un modelo secuencial con tres capas densas. La primera capa tiene 8 unidades y una forma de entrada de 1 dimensión, la segunda capa tiene 4 unidades y la tercera capa tiene una sola unidad. Hemos compilado el modelo con el optimizador Adam y la función de pérdida mean squared error, y luego hemos entrenado el modelo durante 500 épocas utilizando el conjunto de datos xs e ys.

Una vez que hayas entrenado el modelo, puedes utilizarlo para hacer predicciones para valores x más grandes y alejados de los de la secuencia utilizando el método predict. Por ejemplo:

```python
# Hacemos una predicción para un valor x mayor y alejado de los de la secuencia
prediction = model.predict([10])
print(prediction)
```

# Extrapolación

Como podemos comprobar, el modelo puede no ser capaz de hacer predicciones precisas para valores x muy alejados de los de la secuencia de entrenamiento, ya que el modelo sólo ha visto un rango limitado de valores x durante el entrenamiento.

Las redes neuronales suelen ser mejores en la interpolación que en la extrapolación. La interpolación se refiere a hacer predicciones para valores que están dentro del rango de los datos de entrenamiento, mientras que la extrapolación se refiere a hacer predicciones para valores que están fuera del rango de los datos de entrenamiento.

En general, las redes neuronales son buenas en la interpolación debido a que han sido entrenadas con un gran número de ejemplos y, por lo tanto, tienen una buena comprensión de cómo se relacionan los datos de entrada y salida. Sin embargo, cuando se trata de hacer predicciones fuera del rango de los datos de entrenamiento (extrapolación), las redes neuronales suelen tener un rendimiento más bajo debido a que carecen de información sobre cómo se comportan los datos en ese rango.

### Regresión lineal

En esta técnica se busca encontrar la mejor recta que se ajuste a un conjunto de puntos en un plano cartesiano. La recta se representa mediante una ecuación de la forma y = mx + b, donde m es la pendiente de la recta y b es el punto de corte con el eje y. Destacan 3 tipos:

 - Regresión lineal simple: Utiliza el método de mínimos cuadrados para estimar los parámetros de un modelo lineal. El objetivo es minimizar el error: la suma de los cuadrados de las diferencias entre los valores observados y los valores predichos por el modelo.

 - Regresión lineal Ridge: es una técnica de regularización que se utiliza para evitar el sobreajuste en la regresión lineal. Añade un término de penalización al criterio de minimización para reducir la complejidad del modelo y mejorar su capacidad de generalización.

 - Regresión lineal Lasso: similar a Ridge pero no busca generalizar sino seleccionar automáticamente las variables más importantes.







### Regresión cuadrática

Es similar a la regresión lineal, pero en este caso se busca encontrar la mejor parábola que se ajuste a un conjunto de puntos. La parábola se representa mediante una ecuación de la forma y = ax^2 + bx + c, donde a, b y c son parámetros.



### Regresión LOESS

La regresión LOESS (Locally Weighted Scatterplot Smoothing, también conocida como regresión local lineal no uniforme) es una técnica de regresión estadística utilizada para estimar relaciones entre variables. Esta técnica es diferente de la regresión lineal tradicional, ya que permite una estimación no lineal de los datos, lo que significa que la pendiente de la línea de regresión no siempre es constante. En lugar de eso, la regresión LOESS usa 'ventanas' móviles, donde la estimación de la pendiente en una ubicación particular se basa en los datos de los puntos vecinos. Esto permite que la regresión LOESS se ajuste mejor a los datos, dando resultados más precisos y robustos que la regresión lineal tradicional.

### Regresión logística

Es una técnica estadística que se utiliza para modelar la relación entre una variable dependiente categórica y una o varias variables independientes. La variable dependiente es una variable binaria, es decir, solo tiene dos posibles valores, como por ejemplo "sí" o "no", "verdadero" o "falso", o "éxito" o "fracaso". La(s) variable(s) independiente(s) pueden ser cualitativas o cuantitativas.

La regresión logística se basa en la idea de que la probabilidad de que un evento ocurra (por ejemplo, la probabilidad de que un paciente tenga una enfermedad) es una función de las variables independientes (por ejemplo, los síntomas del paciente, su edad, su historial médico, etc.). La regresión logística se utiliza para estimar esta probabilidad y para predecir la probabilidad de que un evento ocurra en el futuro.

La regresión logística se utiliza comúnmente en problemas de clasificación, especialmente en el campo de la medicina, la psicología, la economía, la criminología y las ciencias sociales, para identificar los factores predictivos de un evento, para predecir el riesgo de un evento, y para generar un modelo de clasificación.
